#### 7.copy()
- 04ì¥ listì±•í„°ì—ì„œ ë³´ì•˜ë˜ê²ƒ ì²˜ëŸ¬ ë©¤ë²„í•¨ìˆ˜ copyëŠ” shallow copy ì¦‰ ì–•ì€ ë³µì‚¬ì´ë‹¤

```python
dict01 = {1:'a',2:'b',3:'c',4:[1,2,3]}
dict02 = dict01.copy()
print(dict01 is dict02) # ë‹¤ë¥¸ ì£¼ì†Œì— ì €ì¥ì´ ë˜ê¸´ í•œë‹¤

# í•˜ì§€ë§Œ ë‚´ë¶€ì ìœ¼ë¡œ ì›ì†Œê°€ ê°™ì€ ì£¼ì†Œë¥¼ ê°–ê¸°ë•Œë¬¸ì— ì˜í–¥ì„ ë°›ëŠ”ë‹¤
dict01[4][0]=10
print(dict01)
print(dict02)


import copy
dict01 = {1:'a',2:'b',3:'c',4:[1,2,3]}
dict02 = copy.deepcopy(dict01)
dict01[4][0]=10
print(dict01)
print(dict02) # ì˜í–¥ì„ ë°›ì§€ ì•ŠëŠ”ë‹¤
```

### ğŸ“dict + forë¬¸

```python
# ë°˜ë³µë¬¸ìœ¼ë¡œ valueë§Œ ì¶œë ¥
dict01 = {1:'a',2:'b',3:'c'}
for i in dict01:
    print(dict01.get(i))


for i in dict01.values():
    print(i)
```

### ğŸ“ë”•ì…”ë„ˆë¦¬ í‘œí˜„ì‹
- {ì‹¤í–‰ë¬¸ for í‚¤:ê°’ in ë”•ì…”ë„ˆë¦¬}
- ì£¼ì˜í• ê²ƒì€ ì‹¤í–‰ë¬¸ì•ˆì— í˜•ì‹ì´ í‚¤:ê°’ ì€ ìµœì†Œí•œìœ¼ë¡œ ìœ ì§€í•´ ì£¼ì–´ì•¼í•œë‹¤

```python
dict01 = {1:'a',2:'b',3:'c'}
print({key**2:value for key,value in dict01.items()})
print({key:value+"!" for key,value in dict01.items()})
```

```python
keys = [1,2,3,4,5]
print({key:value for key,value in dict.fromkeys(keys).items()})
print({key+2:value for key,value in dict.fromkeys(keys).items()})
print({key:10 for key in dict.fromkeys(keys)})
```

```python
# ì¡°ê±´ë¬¸ í•¨ê»˜ ì‚¬ìš©ê°€ëŠ¥
# valueê°€ ì§ìˆ˜ì¸ê²ƒë§Œ ê°€ì ¸ì˜¤ì
dict01 = {'a':1,'b':2,'c':3,'d':4}
print({key:value for key,value in dict01.items() if value%2==0})
```

### ğŸ“ì§‘í•©ì—°ì‚°ì
- | (í•©ì§‘í•©)
- & (êµì§‘í•©)
- \- (ì°¨ì§‘í•©)
- ^ (ëŒ€ì¹­ì°¨ì§‘í•©)

```python
set01 ={1,2,3,4}
set02 = {3,4,5,6}
print(set01|set02)
print(set01.union(set02))
```

```python
set01 ={1,2,3,4}
set02 = {3,4,5,6}
print(set01&set02)
print(set01.intersection(set02))
```

```python
set01 ={1,2,3,4}
set02 = {3,4,5,6}
print(set01-set02)
print(set01.difference(set02))
```

```python
set01 ={1,2,3,4}
set02 = {3,4,5,6}
print(set01^set02)
print(set01.symmetric_difference(set02))
```

```python
# í• ë‹¹ì„ ìœ„í•´ì„  ê°’ì„ ìƒˆë¡œìš´ ë³€ìˆ˜ì— ì €ì¥í•˜ë˜ì§€ ê¸°ì¡´ì˜ setì„ ì—…ë°ì´íŠ¸ ì‹œì¼œì£¼ì–´ì•¼ í•œë‹¤
set01 ={1,2,3,4}
set02 = {3,4,5,6}
set01.update(set02)
print(set01)

set01 ={1,2,3,4}
set02 = {3,4,5,6}
set01.intersection_update(set02)
print(set01)

set01 ={1,2,3,4}
set02 = {3,4,5,6}
set01.difference_update(set02)
print(set01)

set01 ={1,2,3,4}
set02 = {3,4,5,6}
set01.symmetric_difference_update(set02)
print(set01)
```

```python
# íŒŒì´ì¬ ìŠ¤ëŸ½ê²Œ í• ë‹¹ë„ ê°€ëŠ¥í•˜ë‹¤
set01 ={1,2,3,4}
set02 = {3,4,5,6}
set01|=set02
print(set01)
```
