#####  <ë³´ì¶©> Slicingì˜ ê·œì¹™ ì‹¬í™”
- [i:j:k]
- iê°€ ì‹œì‘ j-1ê¹Œì§€ Kê°œì”© ê±´ë„ˆê°€ë©°
- i ê°€ ë¹„ì–´ìˆìœ¼ë©´ => i=0
- j ê°€ ë¹„ì–´ìˆìœ¼ë©´ => j=len(list)
- i j ê°€ ìŒìˆ˜ì¼ìˆ˜ ìˆë‹¤
- k ê°€ -1 ì´ë©´ ì—­ìœ¼ë¡œ ì¶œë ¥í•œë‹¤  => ì´ë•ŒëŠ” i j ë„ ì—­ìœ¼ë¡œ ìƒê°í•˜ì

```python
a = list(range(11))
print(a)
print(a[:])  # ì „ë¶€ ë‹¤
print(a[:5]) # ì²˜ìŒë¶€í„° 5ì „ê¹Œì§€, í˜¹ì€ ì²˜ìŒë¶€í„° 5ê°œ
print(a[3:]) # 3ë¶€í„° ëê¹Œì§€
print(a[::2]) # indexê°€ 0 2 4 ..

print(a[-3:]) # j ì— len(a) ë“¤ì–´ê°€ì§„ë‹¤
print(a[:-3]) # i ì— 0ì´ ë“¤ì–´ê°€ì§„ë‹¤

print(a[::-1]) # i = len(a) j =0
print(a[10:4:-1])


print(a[0:10:-1]) # ë°”ë³´ê°™ì€ ì½”ë“œë‹¤ ì¶œë ¥ ê²°ê³¼ ì—†ìŒ
print(a[-8:-3:-2]) # ë°”ë³´ê°™ì€ ì½”ë“œë‹¤ ì¶œë ¥ ê²°ê³¼ ì—†ìŒ
```

### mutable vs immutblae type
- mutableí•œ ê°ì²´ëŠ” ê°’ì„ ìˆ˜ì •(í• ë‹¹) ì‚­ì œ í• ìˆ˜ìˆë‹¤
- immutable í•œ ê°ì²´ëŠ” ê°’ì„ ìˆ˜ì • ì‚­ì œ(í• ë‹¹) í•  ìˆ˜ ì—†ë‹¤
- mutableí•œ ê°ì²´ëŠ” list set dict (set ê³¼ dictëŠ” ë‹¤ìŒì¥ì—ì„œ ë°°ìš¸ê²ƒì´ë‹¤)
- ë‚˜ë¨¸ì§€ëŠ” ë‹¤ immuatableí•˜ë‹¤

#### 1.mutableí•œ ê°ì²´

```python
# ìš°ë¦¬ê°€ ë°°ìš´ sequence ê°ì²´ì¤‘ listëŠ” mutableí•œ ê°ì²´ë‹¤

#indexing í™œìš”í•˜ì—¬ ê°’ì„ ìˆ˜ì •
a = [1,2,3,4]
a[0] = 10
print(a)


#slicing í™œìš”í•˜ì—¬ ê°’ì„ ìˆ˜ì •
a = [1,2,3,4,5,6,7]
a[0:4] = [10,20,30,40] # indexê°€ ìë™ì ìœ¼ë¡œ ì¡°ì ˆëœë‹¤
print(a)

a = []
a[0:4] = [10,20,30,40] # indexê°€ ìë™ì ìœ¼ë¡œ ì¡°ì ˆëœë‹¤
print(a)


# <ì£¼ì˜> : indexingìœ¼ë¡œ ì ‘ê·¼í• ë• index ë²”ìœ„ë¥¼ ê·¸ëŒ€ë¡œ ì§€ì¼œì¤˜ì•¼í•œë‹¤
# ì•„ë‹ˆë©´ index out of range errorë°œìƒ
a = []
a[0] = 10
print(a)
```

```python
# ìš°ë¦¬ê°€ ë°°ìš´ sequence ê°ì²´ì¤‘ listëŠ” mutableí•œ ê°ì²´ë‹¤

#ê°’ì„ ì‚­ì œ í• ìˆ˜ë„ ìˆë‹¤ => del í•¨ìˆ˜ ì´ìš©
a = [1,2,3,4,5,6,7]
del(a[0])
print(a)

del(a[-1:-3:-1]) #ë’¤ì—ì„œ 2ê°œ ì§€ì›€
print(a)
```

#### 2. immuableí•œ ê°ì²´

```python
# ìš°ë¦¬ê°€ ë°°ìš´ sequence ê°ì²´ì¤‘ str range tupleì€ ëª¨ë‘ immutableí•˜ë‹¤

a = 'str'
del(a[0])       # error :'str' object doesn't support item deletion
a[0]=a          # error :'str' object does not support item assignment


a = (1,2,3,4)
del(a[0])        # error :'tuple' object doesn't support item deletion
a[0] = 10
a[0:2] = (1,20) # error :'tuple' object does not support item assignment


```

#### <ë³´ì¶©> immuableí•˜ë‚˜ëŠ” ì˜ë¯¸
- í”„ë¡œê·¸ë˜ë° ì–¸ì–´ë¥¼ ì²˜ìŒ ê³µë¶€í•˜ëŠ” ì‚¬ëŒì´ë¼ë©´ immutableí•˜ë‹¤ëŠ” ì˜ë¯¸ê°€ ë§¤ìš° í˜¼ë€ìŠ¤ëŸ¬ìš¸ìˆ˜ ìˆë‹¤
- ì˜ˆë¥¼ë“¤ì–´ì„œ a='str' ë¥¼ a = 'atr'ë¡œ ë‹¤ì‹œ ì„ ì–¸í•˜ë©´ ê°’ì„ ìˆ˜ì •í•˜ëŠ” ê²ƒì´ ì•„ë‹Œê°€ í•˜ëŠ” ì˜ë¬¸ì´ ë“¤ ìˆ˜ ìˆë‹¤
- ë‹¤ìŒì˜ ë‚´ìš©ì„ ì˜ ì´í•´í•´ë³´ì
    - mutableí•œ ê°ì²´ëŠ” ìˆ˜ì •ì´ ì´ë£¨ì–´ì ¸ë„ ê°ì²´ì˜ ì£¼ì†Œê°’ì´ ë™ì¼í•˜ë‹¤
    - immutableí•œ ê°ì²´ëŠ” ìˆ˜ì •ì´ ë¶ˆê°€ëŠ¥í•˜ê³  a='atr'ê³¼ ê°™ì´ ë³€ìˆ˜ì— ê°’ì„ ë‹¤ì‹œ ì„ ì–¸í•˜ëŠ” ê²ƒì€ ì•„ì˜ˆ ë‹¤ë¥¸ ê°ì²´ê°€ ìƒì„±ë˜ëŠ” ê²ƒì´ë‹¤, ì¦‰ mutableí•˜ë‹¤ëŠ” ì˜ë¯¸ê°€ ì•„ë‹ˆë‹¤

```python
# ëª¨ë“  ê°ì²´ëŠ” ì£¼ì†Œë¥¼ ê°€ì§„ë‹¤

# ë¦¬ìŠ¤íŠ¸ì˜ ì˜ˆì‹œë¥¼ ë³´ì
a = [1,2,3]
print(id(a))
a[0]=4
print('ìˆ˜ì •')
print(id(a)) # ìˆ˜ì • ì „ê³¼ ìˆ˜ì • í›„ì˜ ì£¼ì†Œê°’ì´ ë™ì¼í•˜ë‹¤

print("-"*20)
# stringì˜ ì˜ˆì‹œë¥¼ ë³´ì
a = 'str'
print(id(a))
a = 'atr'
print('ìˆ˜ì •')
print(id(a)) # ìˆ˜ì • ì „ê³¼ ìˆ˜ì • í›„ì˜ ì£¼ì†Œê°’ì´ ë‹¤ë¥´ë‹¤
```
