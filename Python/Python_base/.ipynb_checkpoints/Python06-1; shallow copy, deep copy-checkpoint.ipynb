```python
# object interning (해석시 재사용)
# why? => optimize memory
# interning 이라는 말 자체가 사실은 다른 객체라는것이다

# [-5, 256]
# some strings
# boolean 
# Interning does not apply to other types of objects such as large integers,
# most strings, floats, lists, dictionaries, tuples.
```

```python
# interning 원래는 다른 주소값이다
a = 3
b = 3
a is b
a+=1 # immuable이기때문에 재할당
a is b

str01 = 'abc'
str02 = 'abc'
print(str01 is str02)
str01 = 'abc'+'a'#immutable이기 때문데 재할당
print(str01 is str02)


# interning 적용 x
list01 = [4,5]
list02 = [4,5]
print(list01 is list02)
```

```python
#모든 것이 객체다
# ex) list
# list is collection of objects
# list itself is object , each element is also object

list01 = [1,2,3,[4,5]]
print(id(list01))
print(id(list01[0]))
print(id(list01[1]))
print(id(list01[2]))
print(id(list01[3]))
```

```python
# 리스트 내에서도 interning 존재
list01 = [1,1,3,[4,5],[4,5]]
print(id(list01[0]),id(list01[1])) # interning 적용
print(id(list01[3]),id(list01[4])) # interning 적용안된다 list이니깐
```

```python
# 참조선언 

# immutable 대입
# 애초에 immutable하기 때문에 수정도 불가 => 수정시 다른 주소값이다
a = 3 
b = a
print(a is b)
a+=1 # 재할당
print(a is b)

# mutable 대입 (shallow copy)
# 완전히 같은 주솟값
# 어떤짓을 하던 영향을 받는다
list01 = [1,1,3,[4,5],[4,5]]
list02 = list01 
list01[0] =2 
print(list01,list02)

```

```python
# copy() (shallow copy)
list01 = [1,1,3,[4,5],[4,5]]
list02 = list01.copy()

list01[0] =2 
print(list01,list02) # 아무 일도 일어나지 않는 이유? deep copy여서? 아니다!

list01[3][0]=3
print(list01,list02) # 영향을 받는다

# 도대체 무슨 차이인 것일까? 
list03 = list01.copy()
list04 = list01[:]
import copy
list05 = copy.copy(list01)

print(id(list03),id(list04),id(list05)) # 다 다른 주소의 리스트를 생성한다 
# 하지만 내부적으로 list01의 원소들이 reference 되어있다

print(id(list03[3]),id(list04[3]),id(list04[3]))
# 분명히 [4,5]는 리스트 객체임으로 interning의 대상이 아닌데, 새로 생성된 모든 list안에 
# 4번재 원소가 다 같은 주소를 가르키고 있다
# 그렇기 때문에 id(list03[3])이 가르키는 주소의 할당된 값을 수정하거나 변경하면
# list03 list04 list05가 다 영향을 받을수 밖에 없다

# 그렇다면 아래의 경우에는 왜 아무런 영향을 주지 않았던 걸까?
list01[0] =2 
print(list01,list02)
# 이경우에는 list[0]이 가르키는 주소에 할당된 값을 수정하거나 변경한 경우가 아닌
# list01의 첫번째 요소에 다른 객체가 들어간것이다
# 그렇기 때문에 다른 주소를 갖는 list02에는 전혀 영향을 주지 않는다
```

```python
# 그렇다면 아래의 경우에는 왜 아무런 영향을 주지 않았던 걸까?
list01 = [1,1,3,[4,5],[4,5]]
list02 = list01.copy()
list01[0] =2 
print(list01,list02)

# 이경우에는 list[0]이 가르키는 주소에 할당된 값을 수정하거나 변경한 경우가 아닌
# list01의 첫번째 요소에 다른 객체가 들어간것이다
# 그렇기 때문에 다른 주소를 갖는 list02에는 전혀 영향을 주지 않는다
```

```python
# 결국 immutable객체는 copy에 대해 신경쓸 필요조차 없다
# 우리가 자주 사용하는 mutable객체인 list dict set에 대해서만 주의를 요하면 된다
# 추가적으로 후에 배울 class를 생성할때도 class instance에 주의가 필요하다

# Deep copy를 하고싶으면
# import copy 후에 copy.deepcopy()사용 
# 영향을 미치지 않는다
```
