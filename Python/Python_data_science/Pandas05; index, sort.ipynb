```python
import numpy as np
import pandas as pd
```

```python
df01 = pd.DataFrame(np.random.randint(0,10,(6,4)))
df01.index = pd.date_range('20210901',periods=6)
df01.columns = ['A','B','C','D']
df01.index.name = 'id'
display(df01)
```

### 인덱스를 재정의하는 방법?
- reset_index(inplace=) : 새로운 인덱스 할당, 기존 인덱스를 열로 추가
- reindex(index=[],columns=[])
    - 완전히 새로운 인덱스를 덮어쓰는게 아니다 => fancy indexing 처럼 동작한다
    - 원래 값이 따라간다
 

```python
# reset_index
df01.reset_index()
```

```python
# reindex
dates  = pd.date_range('20210901',periods=6)
new_row_index = np.random.permutation(dates)
new_col_index = ['B','C','A','D']
df01 = df01.reindex(index=new_row_index,columns=new_col_index)
```

### 정렬
- sort_index(axis=,inplace=)
- sort_values ; 데이터프레임과 시리즈에서 다 사용가능, 인덱스도 함께 움직인다
    - Series일때는 Series.sort_values(ascending=,inplace=)
    - DataFrame 일때는 DataFrame.sort_values(by=컬럼,ascending)
    - 컬럼이 여러개일경우 list형식
    - 컬럼에 row index도 사용가능하다 단, index.name이 있어야한다

```python
# sort_index
display(df01.sort_index(axis=0,ascending=False))
```

```python
# sort_index
display(df01.sort_index(axis=1))
```

```python
# sort_index
display(df01.sort_index(axis=1).sort_index(axis=0))
```

```python
# sort_values
# 반드시 label로 부른다
df01.sort_values(by='A')
```

```python
# sort_values
# 두개 이상 열로도 가능
df01.sort_values(by=['A','B'])
```

```python
# sort_values
# index.name 도 사용가능하다
df01.sort_values(by=['A','id'])
```

#### Series에서도 사실 가능하다
- 다만 sort_values()에서 by가 필요없다 어짜피 1차원 데이터

```python
s01 = pd.Series([5,4,3,2,1],index =[1,2,3,4,5])
s01.sort_index(inplace=True,ascending=False)
print(s01)
```

```python
s01 = pd.Series(np.random.randint(1,10,5))
print(s01)
s01.sort_values(inplace=True)
print(s01)
```
