### 난수생성 
- np.random의 subpackages들을 이용해서 난수 발생
    - np.random.rand(갯수) : uniform(0,1] parameter 고정이다
    - np.random.uniform(start,end,size=()) : uniform[start,end)
    - np.random.randint(start,end,size=()) : uniform[start,end) 
    - np.random.randn(갯수) : N(0,1) parameter 고정이다
    - np.random.normal(mean,sd,size=()) : N(mena,sd)

```python
print(np.random.rand(10)) 
print(np.random.rand(2,3))  # 괄호 하나 더하면 에러다
```

```python
print(np.random.randn(10))
print(np.random.randn(3,3))
```

```python
print(np.random.randint(3))
print(np.random.randint(3,10,10))
print(np.random.randint(3,10,size=(3,3)))
```

### samling
- 주어진 1차원 배열 혹은 정수로 부터 랜덤 추출
- 주어진 1차원 ndarray로 부터 랜덤으로 샘플링
- 정수가 주어진 경우, np.arange(해당숫자)로 간주
- np.random.choice(array,size,replace,p)

```python
print(np.random.choice(5,10))
print(np.random.choice(5,size=(2,5)))
```

```python
# 확률 옵션 가능
print(np.random.choice(5,10,p=[0.1,0.2,0.3,0.2,0.2]))
```

```python
# 주사위100번 던지는 실험
# 중복 replace 옵션 가능
np.random.choice(np.arange(1,7),100,replace=True)
```

### counting
- np.unique (정수일 경우에 유리하겠지)
    - 중복된 값을 제거한 새로운 배열을 return
    - return_counts 옵션 사용 가능 => Tuple로 새로운 배열과 카운트 배열 두개 넘겨준다

- np.bincount
    - 특정 범위 내의 카운트를 집계 가능

```python
# np.unique()
random_int = np.random.randint(1,10,10)
print(np.unique(random_int))
```

```python
random_int = np.random.randint(1,10,10)
print(random_int)
new_int, count = np.unique(random_int,return_counts=True)
print(new_int)
print(count)
```

```python
# np.bincount()
random_int = np.random.randint(1,20,10)
print(random_int)
np.bincount(random_int,minlength=21)
```
