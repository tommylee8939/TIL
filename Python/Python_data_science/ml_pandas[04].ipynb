#### 프레임 병합
- pd.merge  : 공통의 열을 기준으로 병합 -> inner join , 인덱스 병합도 가능하다
- concat    : 연결
- frm.join  : 인덱스를 기준으로 병합

```python
data1 = {
    '학번' : [1,2,3,4] , 
    '이름' : ['려명' , '승윤' , '태형' , '지엽'] , 
    '학년' : [2,4,1,3]
}
data2 = {
    '학번' : [1,2,4,5] , 
    '학과' : ['CS' , 'AI' , 'AI' , 'CS'] , 
    '학점' : [2.4 , 4.5 , 1.7 , 3.9]
}
```

```python
stu_frm   = pd.DataFrame(data1)
major_frm = pd.DataFrame(data2)
display(stu_frm)
print()
display(major_frm)
```

```python
pd.merge(stu_frm , major_frm , how='inner')
```

```python
pd.merge(stu_frm , major_frm , how='outer')
```

```python
pd.merge(stu_frm , major_frm , how='left')
```

```python
pd.merge(stu_frm , major_frm , how='right')
```

```python
import seaborn as sns

titanic_datasets = sns.load_dataset('titanic')
iris_datasets    = sns.load_dataset('iris')
```

```python
titanic_datasets.info()
```

```python
# titanic_subset_frm = titanic_datasets[['age', 'sex', 'class', 'fare', 'survived']]
titanic_subset_frm = titanic_datasets.loc[ : , ['age', 'sex', 'class', 'fare', 'survived']]
titanic_subset_frm.head()
```

```python
titanic_subset_frm.describe()
```

```python
print('승객수 - ' , len(titanic_subset_frm) )
```

```python
print('선실등급에 따른 그룹 - ')
class_grp = titanic_subset_frm.groupby(['class'])

for key , group in class_grp :
    print('key - ' , key)
    print('len - ' , len(group))
    print()
    print(group)
    print("*" * 50)
    
```

```python
class_grp.mean()
```

```python
class_grp.sum()
```

```python
class_grp.agg([np.mean , np.sum])
```

```python
class_grp_third = class_grp.get_group('Third')
class_grp_third
```

```python
print('선실등급에 성별에 따른 그룹 - ')
class_sex_grp = titanic_subset_frm.groupby(['class' , 'sex'])

for key , group in class_sex_grp :
    print('key - ' , key)
    print('len - ' , len(group))
    print()
    print(group)
    print("*" * 50)
```

```python
class_sex_grp.get_group( ('First', 'female') )
```

```python
iris_species_grp = iris_datasets.groupby(['species'])
for key , group in iris_species_grp :
    print('key - ' , key)
    print('len - ' , len(group))
    print()
    print(group)
    print("*" * 50)
```

```python
print('[문제] iris 종별(species) 가장 큰 값과 가장 작은 값의 비율을 구한다면? - ')

def get_ration(x) :
    return x.max() / x.min()

iris_species_grp.agg(get_ration)
# iris_species_grp.max() / iris_species_grp.min() 
```

```python
print('[문제] iris 종별(species) 가장 큰 petal_length 3개를 구한다면? - ')

# iris_datasets.sort_values(by='petal_length' , ascending=False)[:3]

def petal_length_fun(frm) :
    return frm.sort_values(by='petal_length' , ascending=False)[:3]

iris_species_grp.apply(petal_length_fun)

# iris_datasets.sort_values(by='petal_length' , ascending=False).groupby('species').head(3)
```
