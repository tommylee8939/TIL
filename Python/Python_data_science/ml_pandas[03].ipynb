```python
def seriesInfo(ary) :
    print('series - index + values \n' , ary , type(ary))
    print('index - ', type(ary.index)  , ary.index)
    print('value - ', type(ary.values) , ary.values)

def frmInfo(df) :
    display('DataFrame - ' , df)
    print('shape    - ' , df.shape )
    print('size     - ' , df.size)
    print('ndim     - ' , df.ndim)
    print('index    - ' , df.index   , type(df.index))
    print('columns  - ' , df.columns , type(df.columns))
    print('values   - \n' , df.values  , type(df.values))
```

```python
from bs4 import BeautifulSoup
import requests 
import re
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds')
print('response - ' , response)
```

```python
# response.text
soup = BeautifulSoup(response.text , 'lxml')
rows = soup.select('div > ul > li')
print('type - ' , type(rows))
print()

etfs = {}
etf_name   = []
etf_market = []
etf_ticker = []
for row in rows : 
    # print(row.text)
    # print("*"* 100)
    try : 
        etf_name = re.findall('^(.*) \(NYSE' , row.text)
        # print(etf_name)
        etf_market = re.findall('\((.*)\|' , row.text)
        # print(etf_market)
        etf_ticker = re.findall('NYSE Arca\|(.*)\)' , row.text)
        # print(etf_ticker)
        if ( (len(etf_name) > 0 ) & (len(etf_market) > 0) & (len(etf_ticker) > 0) ) :
            etfs[ etf_ticker[0] ] = [ etf_market[0] , etf_name[0]]
    except AttributeError as err :
        pass
# print(len(etf_name) , len(etf_market) , len(etf_ticker))    
```

```python
temp_frm = pd.DataFrame(etfs)
temp_frm
```

```python
import seaborn as sns
datasets = sns.load_dataset('titanic')
print('type - ')
print( type(datasets))
```

```python
datasets.head()
```

```python
print('age를 기준으로 성인과 미성년자를 구별하는 열을 추가한다면? - ')
print()
datasets['adult/chile'] = datasets.apply(lambda frm :  'adult' if frm['age'] >= 20 else 'child' , axis = 1)
```

```python
datasets.head()
```

```python
print('category 열을 만들어서 age를 기준으로 20살이 넘으면 성별을 그대로 사용하고 미만이면 성별관없이 child 로 ? - ')
datasets['category'] = datasets.apply(lambda frm :  frm['sex'] if frm['age'] >= 20 else 'child' , axis = 1)
```

```python
datasets.head()
```

```python
print('embark_town의 문자열 갯수를 별도의 컬럼인 embark_len 으로 추가하고 싶다면? - ')
print()
datasets['embark_len'] = datasets['embark_town'].apply(lambda frm :  len(str(frm) )  )
```

```python
datasets[['embark_town' , 'embark_len']].head()
```

```python
print('15세 이하는 child ,  15~60 까지는 adult , 61 이상은 elderly 로 분류를해서 age_division 컬럼에 저장 - ')
print('lambda 식으로 apply함수를 이용해서 구현한다면 - ')
print()

datasets['age_division'] = datasets.apply(lambda frm : 'child' if frm['age'] <= 15 else 'adult' if frm['age'] <= 60 else 'elderly' ,  axis = 1 )

```

```python
condition01 = lambda frm : frm['sex'] if frm['age'] >= 20 else 'child'
condition02 = lambda frm : "I don't know" if pd.isnull(frm['age']) else condition01(frm)

datasets['age/gender'] = datasets.apply(condition02 , axis = 1) 
```

```python
data = {
    '국어' : [50, 65, 87, 10, 100, 89, 32, 83, 100, 58],
    '영어' : [100, 30, 20, 11, 88, 97, 100, 100, 10, 57],
    '수학' : [100, 59, 88, 29, 99, 29, 22, 99, 29, 50]
}
```

```python
# 10명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.
# 1. 
# 학생 이름을 나타내는 열을 포함시키지 않고 
# 데이터프레임 df_score1 을 생성한 후, 
# df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
# reset_index 명령으로 이 인덱스 열을 명령으로 
# 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.

# 2.
# 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 
# set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.
```
