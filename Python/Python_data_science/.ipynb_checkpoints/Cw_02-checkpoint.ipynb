```python
# 연습
try : 
    html = urlopen('https://www.imdb.com/title/tt4154756/reviews?ref_=tt_ql_3')
    #print(html.read())
except HTTPError as he:
    print('http error')
except URLError as ue:
    print('url error')
else:
    soup = BeautifulSoup(html.read(),'html.parser')
```

```python
soup.select('.lister-item-content')[0]
```

```python
# 웬만하면 list comprehension 쓰는게 좋을거 같다

titles  = list(map(lambda x:x.text.strip(),soup.select('.title')))
points =[i.find(class_='rating-other-user-rating').text.strip() if i.find(class_='rating-other-user-rating') else None for i in soup.select('.lister-item-content')]
writers = list(map(lambda x:x.text.strip(),soup.select('.display-name-link')))
dates =list(map(lambda x:x.text.strip(),soup.select('.review-date')))
contents = list(map(lambda x:x.text.strip(),soup.select('.content')))

pd.DataFrame({'title':titles,'point':points,'writer':writers,'date':dates,'content':contents})
```

```python
titles =[]
names = []
dates =[]
contents =[]
ratings =[]
for item in soup.select('.lister-item-content'):
    titles.append(item.find(class_='title').text.strip())
    names.append(item.find(class_='display-name-link').text)
    dates.append(item.find(class_='review-date').text)
    contents.append(item.find(class_='content').text)    
    if item.find(class_='rating-other-user-rating') is None:
        ratings.append(None)
    else:
        ratings.append(item.find(class_='rating-other-user-rating').text.strip())
        
pd.DataFrame({'title':titles,'rating':ratings,'name':names,'date':dates,'content':contents})
```

### 이미지 로컬에 저장하기

```python
# img 폴더를 생성해보자
import os

img_dir = './img'

if not os.path.isdir(img_dir):
    os.mkdir(img_dir)
```

```python
from selenium import webdriver
import time
driver = webdriver.Chrome('./driver/chromedriver')
driver.get('https://m.search.daum.net/search?w=img&nil_search=btn&DA=NTB&enc=utf8&q=%EC%8B%9C%EC%B6%94')
imgs = driver.find_elements_by_css_selector('.thumb_img')
img_url = []
for img in imgs:
    url = img.get_attribute('src')
    img_url.append(url)
    
time.sleep(5)
driver.quit()
display(img_url)
```

```python
# enumerate 사용해서 img의 파일명으로 사용할 index 생성
for idx,link in  enumerate(img_url):
    #print(idx,link)
    urlretrieve(link,f'./img/{idx}.png') # jpg로도 가능하다 
```

#### CRUD

- insert_one() => shell기반 insertOne()
- insert_many() => shell기반 insertMany()

```python
document01 = {
    'author' : 'kanghee',
    'subject' : 'mongodb',
    'int_value' : 10,
    'list_value' : [1,2,3]
}

document02 = {
    'author' : 'kanghee',
    'subject' : 'mongodb',
    'int_value' : 8,
    'dict_value' : {'a':1,'b':2}
}

document03 = {
    'author' : 'kanghee',
    'subject' : 'mongodb',
    'int_value' : 8,
    'dict_value' : {'a':1,'b':2}
}

```

```python
collection.insert_one(document01)
```

```python
collection.insert_one(document03)
```

```python
collection.insert_many([document01,document02])
```

```python
print(collection.find_one())

# dict로 받아온다
print(type(collection.find_one()))
```
