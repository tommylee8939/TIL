```python
import numpy as np
import pandas as pd
```

```python
def seriesinfo(ary):
    print('data \n',ary)
    print('index:',type(ary.index),ary.index)
    print('value:',type(ary.values),ary.values)
def dfinfo(df):
    display(df)
    print('shape:',df.shape)
    print('size:',df.size)
    print('ndim:',df.ndim)
    print('index:',df.index)     # row index
    print('columns:',df.columns) # column inex 
    print('values:\n',df.values,type(df.values)) #ndarray 형식으로 나온다
```

### 데이터 프레임에서 인덱스, 데이터 조작
    - loc  : 라벨 기반의 2차원 인덱싱 슬라이싱
    - iloc : 기본 정수 인덱스 기반의 2차원 인덱싱 슬라이싱
- loc iloc는 기본적으로 row indexing이다
- loc iloc는 2차원 인덱싱도 가능하다 => 이때 return 타입에 유념

```python
sample_df = pd.DataFrame(np.arange(10,22).reshape(3,4),
                        index = ['a','b','c'],
                        columns = ['col01','col02','col03','col04'])
dfinfo(sample_df)
```

```python
# column indexing
sample_df['col01']
```

```python
# loc iloc은 기본적으로 row indexing이다
sample_df.loc['a']
```

```python
# 마지막 행
sample_df.iloc[-1]
```

```python
# fancy indexing
# 차원이 유지되기 때문에 DataFrame이다

display(sample_df.loc[['a','c']])
display(sample_df.iloc[[0,2]])
```

```python
# boolean indexing

sample_df['col01']>13
sample_df.loc[sample_df['col01']>13]
```

```python
# 2차원 
print(sample_df.loc['a','col01'])
```

```python
print(sample_df.loc['a':'b','col01'])
```

```python
print(sample_df.loc['a',:])
```

```python
print(sample_df.loc[:,'col01'])
print(sample_df.iloc[:,0])
```

```python
print(sample_df.iloc[:,-2:])
```

```python
# 행 추가
sample_df.loc['d'] = 0
print(sample_df)
```

```python
# 열 추가
sample_df['col05'] = sample_df['col04']+sample_df['col03']
```

#### 정렬
- sort_index
- sort_values ; 데이터프레임과 시리즈에서 다 사용가능, 인덱스도 함께 움직인다
    - Series일때는 Series.sort_values(ascending=,inplace=)
    - DataFrame 일때는 DataFrame.sort_values(by=컬럼,ascending)
    - 컬럼이 여러개일경우 list형식
    - 컬럼에 row index도 사용가능하다 단, index.name이 있어야한다

```python
df01 = pd.DataFrame(np.random.randint(0,10,(6,4)))
df01.index = pd.date_range('20210901',periods=6)
df01.columns = ['A','B','C','D']
df01.index.name = 'id'
```

```python
# index를 기준으로 shuffle해보자
display(df01)
random_date = np.random.permutation(df01.index)
df01 = df01.reindex(index = random_date,columns = ['B','A','D','C'])
df01
```

```python
# sort_index
df01.sort_index(axis=0,ascending=False)
df01.sort_index(axis=1).sort_index(axis=0)
df01.sort_index(axis=0).sort_index(axis=1)
```

```python
# 컬럼에 값을 기준으로 행을 정렬한다
# 이때 컬럼값에는 index도 사용할수있다 단 index.name이 있을때
# index도 같이 움직인다
df01.sort_values(by='A')
df01.sort_values(by='D')
```

```python
df01.sort_values(by=['B','id'])
```

```python
df01.sort_values(by=['B','C'])
```

```python
## 타이타닉에서 승객에 대해서 성별의 인원수, 나이별 인원수
## 각각에 대해 sort_values

datasets['sex'].value_counts().sort_values()

datasets['age'].value_counts().sort_values()
```

```python
# 타이타닉에서 승객의 나이를 기준으로 내림차순정렬을 하고 순위를 부여한다면?
rank = datasets.sort_values(by='age')['age'].rank()
datasets['rank'] = rank
```

```python
# 무슨 차이일까?
datasets.sort_values(by='age')['age'].rank()
datasets['age'].rank()
```
