### Scraping

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
# 통신
response = requests.get('https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds')
response
```

```python
# html text
# 하나의 문자열이다 => 이거 어케 parsing할건데 => BeautifulSoup
response.text
```

```python
soup = BeautifulSoup(response.text,'lxml')
soup
```

```python
soup.select('div>ul>li')
```

```python
# response.text
soup = BeautifulSoup(response.text , 'lxml')
rows = soup.select('div > ul > li')
print('type - ' , type(rows))
print()

etfs = {}
etf_name   = []
etf_market = []
etf_ticker = []
for row in rows : 
    # print(row.text)
    # print("*"* 100)
    try : 
        etf_name = re.findall('^(.*) \(NYSE' , row.text)
        # print(etf_name)
        etf_market = re.findall('\((.*)\|' , row.text)
        # print(etf_market)
        etf_ticker = re.findall('NYSE Arca\|(.*)\)' , row.text)
        # print(etf_ticker)
        if (len(etf_name) > 0 ) & (len(etf_market) > 0 ) & (len(etf_ticker) > 0 ) :
            etfs[ etf_ticker[0] ] = [ etf_name[0],etf_market[0]]
    except AttributeError as err :
        pass
```

```python
# response.text
soup = BeautifulSoup(response.text , 'lxml')
rows = soup.select('div > ul > li')
print('type - ' , type(rows))
print()

df01 = pd.DataFrame(columns=['name','market'])
for row in rows : 
    # print(row.text)
    # print("*"* 100)
    try : 
        etf_name = re.findall('^(.*) \(NYSE' , row.text)
        # print(etf_name)
        etf_market = re.findall('\((.*)\|' , row.text)
        # print(etf_market)
        etf_ticker = re.findall('NYSE Arca\|(.*)\)' , row.text)
        # print(etf_ticker)
        if (len(etf_name) > 0 ) & (len(etf_market) > 0 ) & (len(etf_ticker) > 0 ) :
            df01.loc[etf_ticker[0]] = pd.Series([etf_name[0],etf_market[0]],index=['name','market'])
    except AttributeError as err :
        pass
df01
```

### google geocoding api

```python
import googlemaps
```

```python
maps = googlemaps.Client(key='key')
places = ['서울시청','해운대해수욕장','국립현충원']

lat = []
lng = []

for place in places:
    #print('place-',place)
    # print('geocode',maps.geocode(place))
    location = maps.geocode(place)[0].get('geometry')
    lat.append(location['location']['lat'])
    lng.append(location['location']['lng'])

print(lat)
print(lng)
df01 = pd.DataFrame({'위도':lat,'경도':lng},index=places)
df01
```

### 데이터 프레임 행 열 단위의 데이터 가공
- apply 
- lambda식 활용

```python
# 람다식 형식 오랜만이징?
(lambda x : x**2)(2)
```

```python
np.random.seed(100)
df01 = pd.DataFrame(np.random.randint(0,10,(6,4)))
df01.columns = ['A','B','C','D']
df01.index = pd.date_range('20210902',periods=6)
df01
```

```python
df01.apply(max,axis=1) - df01.apply(min,axis=1)
```

```python
df01['E']=df01.apply(lambda x:max(x)-min(x),axis=1)
df01
```

```python
# 이게 좀더 정확한 표현인거 같기는 하다
df01.apply(lambda x:x.max()-x.min(),axis=1)
```

```python
import seaborn as sns
datasets = sns.load_dataset('titanic')
print(type(datasets))
datasets
```

```python
datasets.apply(lambda x: '성인' if x['age']>19  else '아동',axis=1)
```

```python
# 이렇게 해도 된다
datasets['age'].apply(lambda x: '성인' if x>19  else '아동')
```

```python
datasets.apply(lambda x: x['sex'] if x['age']>20 else 'child',axis=1)
```

```python
# embarked town의 문자열 갯수를 별도의 컬림으로
datasets.apply(lambda x: len(str(x['embark_town'])),axis=1)
```

```python
datasets['embark_town'].apply(lambda x: len(str(x)))
```

- 15세 이하는 child ,  15~60 까지는 adult , 61 이상은 elderly 로 분류를해서 age_division 컬럼에 저장
- lambda 식으로 apply함수를 이용해서 구현한다면

```python
datasets['age_division']=datasets.apply(lambda x: 'child' if x['age']<=15 else ('adult' if x['age']<=60 else 'elderly'),axis=1)
```

```python
datasets['age_division'].value_counts()
```

- 승객에 대한 나이(age)와 성별(sex)에 따른 카테고리를 'age/gender' 추가한다
- 조건 1 : 20살이 넘으면 성별을 그대로 사용하고
- 조건 2 : 20살 미만이면 성별에 관계없이 'child' 라고 정의하고 결측값이 있는 경우는 'i don't know'정의한다

```python
datasets.apply(lambda x: x['sex'] if x['age']>20 else 'dont know' if x['age']==None else 'child',axis=1)
```
