> 주의! : ***driver의 url을 옮기면 기존 driver에서 find한 값은 없어진다***

```python
# driver의 url을 옮기면 기존 driver에서 find한 값은 없어진다
path = './driver/chromedriver'
driver = webdriver.Chrome(path)
url = f'https://www.ildong.com/kor/product/list.id?page=1&halt=&prdDisease=&prdCategory=&searchVal=&searchOption=0'
driver.get(url)
prList = driver.find_element_by_class_name('prList')
a_tags = prList.find_elements_by_tag_name('a')
print(a_tags[1].get_attribute('outerHTML'))
a_tags[0].click()
driver.back()

print(a_tags) # a_tags 자체는 다 있을시 몰라도

print(a_tags[1].get_attribute('outerHTML')) # 다시 받아줘야한다
a_tags[1].click()  # 다시 받아줘야한다

driver.close()
```

```python
# css
path = './driver/chromedriver'
driver = webdriver.Chrome(path)
for page in range(1,3):
    url = f'https://www.ildong.com/kor/product/list.id?page={page}&halt=&prdDisease=&prdCategory=&searchVal=&searchOption=0'
    driver.get(url)
    print('page:',page)
    time.sleep(2)
    #print(a_tags)
   
    for product in range(0,9):
        prList = driver.find_element_by_class_name('prList')
        a_tags = prList.find_elements_by_tag_name('a')
        a_tags[product].click()
        tag = driver.find_element_by_class_name('detailCnt2')
        print(tag.text)
        print('*'*50)
        driver.back()

driver.close()
```

```python
# xpath
path = './driver/chromedriver'
driver = webdriver.Chrome(path)
for page in range(1,3):
    url = f'https://www.ildong.com/kor/product/list.id?page={page}&halt=&prdDisease=&prdCategory=&searchVal=&searchOption=0'
    driver.get(url)
    print('page:',page)
    time.sleep(2)
    
   
    for i in range(1,10):
        li = driver.find_element_by_xpath(f'//*[@id="contents"]/div[2]/div[2]/div[2]/div[1]/ul/li[{i}]')
        a_tag = li.find_element_by_xpath('./a')
        a_tag.click()
        tag = driver.find_element_by_class_name('detailCnt2')
        print(tag.text)
        print('*'*50)       
        time.sleep(2)
        driver.back()
driver.close()
```
