```python
data = pd.read_excel('/Users/iganghui/TIL/Python/Python_data_science/Data/시도별 전출입 인구수.xlsx')
data.head()
```

```python
data = data.fillna(method='ffill')
data.head()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 서브셋을 만들면
mask = (data['전출지별']=='서울특별시') & (data['전입지별']!='서울특별시')
data[mask].drop('전출지별',axis=1).rename(columns={'전입지별':'전입지'}).set_index('전입지')
```

```python
# 서울에서 경기도로만 이동량 시간에 따른 변화 
seoul_frm = data[mask].drop('전출지별',axis=1).rename(columns={'전입지별':'전입지'}).set_index('전입지')

x = seoul_frm.columns.astype(int)
y = seoul_frm.loc['경기도']
plt.figure(figsize=(15,5))
plt.plot(x,y,marker='o',markersize=4)
plt.xticks(x,rotation='vertical')
plt.legend(labels=['서울->경기'],loc='best')
plt.ylim(5000,700000)
plt.show()
plt.close()
```

```python
# 하나의 figure안에 여러개의 plt
seoul_frm = data[mask].drop('전출지별',axis=1).rename(columns={'전입지별':'전입지'}).set_index('전입지')


plt.figure(figsize=(15,10))

x = seoul_frm.columns.astype(int)
y = seoul_frm.loc['충청남도']
plt.plot(x,y,marker='o',markersize=4,label='서울->충청남도')

x = seoul_frm.columns.astype(int)
y = seoul_frm.loc['강원도']
plt.plot(x,y,marker='o',markersize=4,label='서울->강원')

x = list(map(int,range(1970,2018)))
y = seoul_frm.loc['경상북도']
plt.plot(x,y,marker='o',markersize=4,label='서울->경상북도')


plt.xticks(x,rotation='vertical')
plt.legend(loc='best')
plt.show()
plt.close()
```

```python
# subplot 그리는방법
# 그리는 방법 여러개 있다
seoul_frm = data[mask].drop('전출지별',axis=1).rename(columns={'전입지별':'전입지'}).set_index('전입지')

x = seoul_frm.columns.astype(int)
y = seoul_frm.loc['경기도']
fig = plt.figure(figsize=(10,10))
ax01 = fig.add_subplot(2,1,1)
ax02 = fig.add_subplot(2,1,2)

ax01.plot(x,y,marker='o',markersize=4)
ax01.set_xticklabels(x,rotation='vertical')
ax01.legend(labels=['서울->경기'],loc='best')
ax01.set_ylim(5000,700000)


x = seoul_frm.columns.astype(int)
y = seoul_frm.loc['강원도']
ax02.plot(x,y,marker='o',markersize=4)
ax02.set_xticklabels(x,rotation='vertical')
ax02.legend(labels=['서울->경기'],loc='best')

plt.show()

plt.close()
```

# barplot
- plt.bar(X,Y)
- X-> 범주형 데이터

```python
import seaborn as sns
iris_datasets = sns.load_dataset('iris')
titanic_datasets = sns.load_dataset('titanic')
display(titanic_datasets)
```

```python
titanic_datasets.groupby('pclass').
```

```python
# 선상 등급에 따른 생존자 수 시각화
plt.figure(figsize=(15,5))
x = titanic_datasets.groupby('pclass').survived.sum().index
y = titanic_datasets.groupby('pclass').survived.sum().values
plt.bar(x,y)

plt.ylabel('survived')
plt.xlabel('pcalss')
plt.xticks(x)
plt.show()
plt.close()
```

```python
plt.figure(figsize=(15,5))
x = titanic_datasets.groupby('pclass').survived.sum().index
y = titanic_datasets.groupby('pclass').survived.sum().values
plt.barh(x,y)

plt.ylabel('survived')
plt.xlabel('pcalss')
plt.xticks(x)
plt.show()
plt.close()
```

```python
# 충청남도 경상북도 강원도 전라남도
# 2010~2017
# 해당하는 subset 생성
seoul_frm = seoul_frm.loc[['충청남도' ,'경상북도' ,'강원도' ,'전라남도'],'2010':'2017']
seoul_frm
```

```python
t_student_frm = seoul_frm.T
t_student_frm
```

```python
# 데이터 프레임을 이용해 바로 그래프를 그릴수도 있다
plt.figure()
t_student_frm.plot(kind='bar',figsize=(20,10))
plt.show()
plt.close()
```

```python
# 전입인구 수의 함계를 구해서 새로운 합계열을 추가하고
# 합계에 대한 가장 큰값으로 정렬
# 수평막대 그래프로 시각화 한다면

seoul_frm['합계'] = seoul_frm.sum(axis=1)
seoul_frm.sort_values(by='합계',ascending=False,inplace=True)
seoul_frm['합계'].plot(kind='barh')
```

```python
seoul_frm[['합계']]
```

```python
# 종을 기준으로 그룹화하여 각 그룹의 평균을 구해서 시각화 한다면?
# 조건 컬럼명 수정 꽃받침길이 꽃받침넓이 꽃잎길이 꽃잎넓이

iris_datasets.rename(columns={'sepal_length':'꽃받침길이',
                              'sepal_width': '꽃받침넓이',
                              'petal_length' : '꽃잎길이',
                              'petal_width' : '꽃잎넓이'},inplace=True)

# barplot
iris_datasets.groupby('species').mean().plot(kind='bar',figsize=(10,8),rot=0)
iris_datasets.groupby('species').mean().T.plot(kind='bar',figsize=(10,8),rot=0)
```

```python
a = iris_datasets.groupby('species').mean().T
a.index.name = '종류'
a.plot(kind='bar',figsize=(10,8),rot=0)
```
