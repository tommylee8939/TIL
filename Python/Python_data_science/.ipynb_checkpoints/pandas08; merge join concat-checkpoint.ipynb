## pandas08; merge join concat

```python
import numpy as np
import pandas as pd
```

```python
# DataFrame 편하게 만들기 위한 함수를 좀 정의해보자
def make_df(cols,ind):
    data = {c:[str(c)+str(i) for i in ind] for c in  cols}
    return pd.DataFrame(data,ind)
```

### Concat
- 그대로 갖다 붙인다고 생각하며 된다
- 그대로 붙이 고 비어있는곳은 NaN처리된다

```python
df01 = make_df('AB',[1,2])
df02 = make_df('AB',[3,4])
display(df01)
display(df02)
```

```python
pd.concat([df01,df02],axis=0)
```

```python
# index가 중요하다는 사실을 알수있지
pd.concat([df01,df02],axis=1)
```

```python
df01 = make_df('AB',[1,2])
df02 = make_df('AB',[1,2])
pd.concat([df01,df02],axis=1)
```

```python
# index가 반복되는 문제 발생
df01 = make_df('AB',[1,2])
df02 = make_df('CD',[1,2])
pd.concat([df01,df02],axis=0)
```

```python
# index 반복 억제
# verify_integrity = True => 에러 발생
df01 = make_df('AB',[1,2])
df02 = make_df('CD',[1,2])
pd.concat([df01,df02],axis=0,verify_integrity=True)
```

```python
# 아예 새로운 인덱스로 생성
df01 = make_df('AB',[1,2])
df02 = make_df('AB',[3,4])
pd.concat([df01,df02],axis=0,ignore_index=True)
```

```python
# 아예 새로운 인덱스로 생성
df01 = make_df('AB',[1,2])
df02 = make_df('BC',[1,2])
display(pd.concat([df01,df02],axis=0))
display(pd.concat([df01,df02],axis=0,ignore_index=True))
```

```python
df01 = make_df('AB',[1,2])
df02 = make_df('BC',[1,2])
display(pd.concat([df01,df02],axis=0))
display(pd.concat([df01,df02],axis=0,join='inner'))
```

```python
df01 = make_df('AB',[1,2])
df02 = make_df('BC',[2,3])
display(pd.concat([df01,df02],axis=1))
display(pd.concat([df01,df02],axis=1,join='inner'))
```
