```python
import numpy as np
import pandas as pd
```

```python
# 1. 각 파일을 pd.read_csv() 로 읽어들인다
movies_df = pd.read_csv('/Users/iganghui/TIL/Python/Python_data_science/Data/movies.dat',sep='::',header=None)
ratings_df = pd.read_csv('/Users/iganghui/TIL/Python/Python_data_science/Data/user_ratings.dat',sep='::',header=None)
users_df = pd.read_csv('/Users/iganghui/TIL/Python/Python_data_science/Data/users.dat',sep='::',header=None)
```

```python
# 2. 데이터 확인 및 구조분석과 각 데이터 프레임에 헤더를 추가
ratings_df.columns = ['userId', 'movieId', 'rating','timestamp']
movies_df.columns = ['movieId', 'title', 'genres']
users_df = users_df.drop([3,4],axis=1)
users_df.columns = ['userId' ,'gender', 'age']
```

```python
print(ratings_df.head())
print(movies_df.head())
print(users_df.head())
```

```python
# 3. 공통 컬럼을 이용한 데이터 병합
temp = pd.merge(ratings_df,movies_df,on='movieId',how='left')
data = pd.merge(temp,users_df,on='userId',how = 'left')
data.head()
```

```python
# 4. user 평점 갯수 순위 확인(sort)
temp_dataframe = pd.DataFrame({
    'userId' : data['userId'].value_counts().index,
    'user_id_cnt' : data['userId'].value_counts()
})

data = pd.merge(data,temp_dataframe,on='userId',how='left')
data
```

```python
# 5.  각 영화의 성별 평균 평점
data[['movieId','rating','gender']].groupby(['movieId','gender']).mean('rating')
```

```python
# 6.  평점이 220개 이상인 영화만 필터링
data.loc[data['movie_id_count']<=200]
```

```python
# 7. 평점이 200개 이상인 영화들의 성별 평균 평점
temp_dataframe = pd.DataFrame({
    'movieId' : data['movieId'].value_counts().index,
    'movie_id_count' : data['movieId'].value_counts()
})
data = pd.merge(data,temp_dataframe,on='movieId',how='left')
data.loc[data['movie_id_count']>=200,['gender','rating']].groupby('gender').mean('rating')
```

```python
# 8. 여성이 가장 좋아하는 영화 순위
female_movie_rating = data.loc[data['gender']=='F',['movieId','rating']].groupby('movieId').mean('rating')
female_movie_rating['rank'] = female_movie_rating['rating'].rank()
female_movie_rating.sort_values(by='rank',ascending=False)
```

```python
# 9. 남녀간의 평균 평점 차이
data[['gender','rating']].groupby('gender').mean('rating')
```

```python
# 10. 평점의 표준편차가 큰 영화
data[['movieId','rating']].groupby('movieId').std()
```
