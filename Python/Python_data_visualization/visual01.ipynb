```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import matplotlib
matplotlib.get_backend() # 아나콘다로 쥬피터 노트북 실행하면 자동적으로 inline이다
```

```python
%matplotlib inline 
# 위에서 말했던 것처럼 이미 inline이기 때문에 굳이 실행할 필요 없다
# inline 아니면 그래프가 셀 바로 밑에 보일 수가 없다!!
```

```python
# %matplotlib notebook 
# 이건 R에서 항상 봤던거 같이 대화형 창이다
# 이러면 그래프가 셀 바로 밑에 생성 되지 않는다
# 어느곳에서 plt.plot을 하던 이 대화형 창에서 실행이 될거다
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
plt.show()
```

```python
# plot은 엄밀히 말하면 파이썬에서 객체다
x = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x,np.sin(x),'-')
```
