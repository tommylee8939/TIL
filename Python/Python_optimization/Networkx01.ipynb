```python
import networkx as nx
```

### Graph 생성
- 그냥 Graph() 는 무방향성 그래프다

```python
# empty graph 
G = nx.Graph()
G.add_node(1)
nx.draw(G)
```

### 노드 추가하기
- 당연히 동일한 노드는 계속 추가해도 추가되지 않음

```python
G = nx.Graph()
G.add_node(1)
G.add_node(1)
print(G.nodes)
print(G.number_of_nodes())
nx.draw(G)
```

```python
# node 여러개 한번에 추가하는것도 당연히 가능하다
# multiple nodes from iterable object
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3,4])

# nodes attributes
G.add_nodes_from([(5,{'color':'red'})])

nx.draw(G, with_labels = True)
```

### Edge 추가하기

```python
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3,4])
G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)
nx.draw(G,with_labels=True)
```

```python
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3,4])
G.add_edges_from([(2,3),(3,4)])
nx.draw(G,with_labels=True)
```

```python
# 사실 노드가 필요 없을수도 있다
# edge를 그리면 당연히 노드도 생성되겠지
G = nx.Graph()
G.add_edges_from([(2,3),(3,4),(1,5)])
nx.draw(G,with_labels=True)
```

### 그래프를 노드로 추가할수도 있다

```python
# node를 추가하거나 edge를 추가하는 방법에는 여러가지가 있을거다
H = nx.path_graph(10)
nx.draw(H)
H.nodes
```

```python
G = nx.Graph()
G.add_nodes_from([1,2,3])
G.add_edge(1,2)
G.add_node(H) # 그래프 자체가 노드로 추가될수도 있다, 사실 어떤 의미인지는 잘 모르겠다
nx.draw(G,with_labels=True)
```

```python
# 그래프 H의 element들을 이용할수도 있다
print(H.edges) # 튜플 형태가 iterable 객체안에 담겨있는걸 볼수있다
G = nx.Graph()
G.add_edges_from(H.edges)
nx.draw(G)
```

### Elements of Nodes

```python
G = nx.Graph()
G.add_nodes_from([1,2,3])
G.add_edge(1,2)
nx.draw(G,with_labels=True)

print(list(G.nodes)) 
print(list(G.edges))
print(list(G.adj[1])) 
print(G.degree[1])
```

```python
# 특정 노드를 더 specify도 가능하다
print(G.edges([2]))   # 2와 연결된 edge들
print(G.edges([2,3])) 
G.degree([2,3])
```

### removing elements

```python
# 문법은 생성할때랑 비슷하다
G = nx.Graph()
G.add_edges_from([(1,2),(2,3),(4,3),(2,7)])
nx.draw(G,with_labels=True)
```

```python
G = nx.Graph()
G.add_edges_from([(1,2),(2,3),(4,3),(2,7)])
G.remove_node(4) # edge도 자연스럽게 없어진다
nx.draw(G,with_labels=True)
```

```python
G = nx.Graph()
G.add_edges_from([(1,2),(2,3),(4,3),(2,7)])
G.remove_edges_from([(2,3),(4,3)]) # node는 남아있다
nx.draw(G,with_labels=True)
```
