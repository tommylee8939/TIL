### Decision Tree 파라미터

- criterion : 분할 품질을 측정하는 기능 (default : gini)
- splitter : 각 노드에서 분할을 선택하는 데 사용되는 전략 (default : best)
- max_depth : 트리의 최대 깊이 (값이 클수록 모델의 복잡도가 올라간다.)
- min_samples_split : 자식 노드를 분할하는데 필요한 최소 샘플 수 (default : 2)
- min_samples_leaf : 리프 노드에 있어야 할 최소 샘플 수 (default : 1)
- min_weight_fraction_leaf : min_sample_leaf와 같지만 가중치가 부여된 샘플 수에서의 비율
- max_features : 각 노드에서 분할에 사용할 특징의 최대 수
- random_state : 난수 seed 설정
- max_leaf_nodes : 리프 노드의 최대수
- min_impurity_decrease : 최소 불순도
- min_impurity_split : 나무 성장을 멈추기 위한 임계치
- class_weight : 클래스 가중치
- presort : 데이터 정렬 필요 여부

```python
iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data  , 
                                                    iris.target ,
                                                    test_size = 0.2 , 
                                                    random_state = 200 ) 

dtc = DecisionTreeClassifier() 
dtc.fit(X_train , y_train)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(dtc, 
                out_file='iris_tree.dot',
                class_names=iris.target_names,
                feature_names=iris.feature_names,
                impurity=True,filled=True)
```

```python
with open("iris_tree.dot") as f:
    dot_graph = f.read() 
graphviz.Source(dot_graph)
```

```python
gscv_dtc = DecisionTreeClassifier()

param = {
    'criterion' : ['gini','entropy'],
    'splitter'  : ['best','random'],
    'max_depth' : [1,2,3],
    'min_samples_split' : [2,3]
}
```

```python
grid_gscv_tree = GridSearchCV(gscv_dtc,
                              param_grid=param,
                              cv=3,
                              refit=True) # 최적의 하이퍼 파라미터를 찾아서 재학습
```

```python
grid_gscv_tree.fit(X_train,y_train)
grid_gscv_tree.cv_results_
```

```python
score_df = pd.DataFrame(grid_gscv_tree.cv_results_)
score_df
```

```python
score_df[['params','rank_test_score']]
```

```python
print(grid_gscv_tree.best_params_)
print(grid_gscv_tree.best_score_)
```

```python
estimator = grid_gscv_tree.best_estimator_
y_predict = estimator.predict(X_test)
print(y_test)
print(y_predict)
print(accuracy_score(y_test,y_predict))
```

### 데이터 전처리
- data encoding : ml에서는 문자열의 값을 입력으로 허용하지 않는다
    - 레이블 인코딩, 원핫 인코딩
    - <주의!> 회귀 알고리즘에서는 적용 하지 않는다 트리계열 알고리즘에서 사용
    - <주의!> 트리 알고리즘에서도 성능이 떨어질수 있다
- 결측값 허용하지 않는다
- feature scailing
    - 정규화(0과 1 사이의 값으로 변환-MinMaxScalar) 
    - 표준화(정규분포를 따르는 형식으로 평균이 0 분산 1 가우시안분포 변환-StandardScalar)

#### data encoding

```python
# 레이블 인코딩
label_items = ['낫또','집밥','ABC쥬스','제육볶음','낫또']
encoder = LabelEncoder()
encoder.fit(label_items)
print('label',encoder.transform(label_items))
print('인코딩 클래스',encoder.classes_)
labels = encoder.transform(label_items)
print('디코딩 복원값',encoder.inverse_transform(labels))
```

```python
# One-hot encoding
label_items = ['낫또','집밥','ABC쥬스','제육볶음','낫또']
encoder = LabelEncoder()
digit_labels = encoder.fit(label_items).transform(label_items)

# digit으로 변환을 먼저 거쳐야한다
print(type(digit_labels))
print(digit_labels.reshape(-1,1))
digit_labels = digit_labels.reshape(-1,1)

one_hot_encoder = OneHotEncoder()
one_hot_encoder_label = one_hot_encoder.fit(digit_labels).transform(digit_labels)
print(one_hot_encoder_label.toarray())
```

```python
# pandas 이용해서 one hot encoding 가능
encoder_df = pd.DataFrame({
    'label_items' : ['낫또','집밥','ABC쥬스','제육볶음','낫또']
})
print(encoder_df)
print(pd.get_dummies(encoder_df))
```

#### feature scaling
- 정규화(0과 1 사이의 값으로 변환-MinMaxScalar) 
    > (x - min)/(max-min) but when the value is negative then we use 1
- 표준화(정규분포를 따르는 형식으로 평균이 0 분산 1 가우시안분포 변환-StandardScalar)
    >  (x - mean(x))/sd(x)

```python
# 아이리스 데이터를 가지고 featuer scaling 해보자
iris_df = pd.DataFrame(data=iris.data,
                       columns=iris.feature_names)
iris_df
```

```python
print('mean')
print(iris_df.mean())

print('std')
print(iris_df.std())

print('var')
print(iris_df.var())
```

```python
# 표준화
std_scaler = StandardScaler()
std_scaler.fit(iris_df)
std_iris = std_scaler.transform(iris_df)

# std_iris = std_scaler.fit_transform(iris_df) # 지양하자
std_iris
std_iris_df = pd.DataFrame(std_iris,columns=iris.feature_names)
std_iris_df
```

```python
# 0과 1사이 값으로 정규화
m_scaler = MinMaxScaler()
m_scaler.fit(iris_df)
m_iris = m_scaler.transform(iris_df)
m_iris
m_iris_df = pd.DataFrame(m_iris,columns=iris.feature_names)
m_iris_df
```

#### <주의!> 스케일링 변환시 주의사항

```python
train_data = np.arange(0,11).reshape(-1,1)
test_data = np.arange(0,6).reshape(-1,1)
print(train_data)
print(test_data)
```

```python
# 학습데이터 정규화
scaler = MinMaxScaler()
scaler.fit(train_data)
m_train_data = scaler.transform(train_data)
print('트레인 데이터',train_data.reshape(-1))
print('정규화 데이터',m_train_data.reshape(-1))
```

```python
# 테스트데이터정규화
# 테스트 데이터의 스케일링 기준은 학습 데이터의 스케일링 기준을 잡아야 한다
# 학습기는 학습 데이터를 기준으로  학습을 진행하기 때문
scaler = MinMaxScaler()
scaler.fit(test_data)    # test_data를 기준으로 fit을 하면 안된다는 거다
m_test_data = scaler.transform(test_data)
print('테스트 데이터',test_data.reshape(-1))
print('정규화 데이터',m_test_data.reshape(-1))
```

```python
# 정리 
scaler = MinMaxScaler()
scaler.fit(train_data)
m_train_data = scaler.transform(train_data)
m_test_data = scaler.transform(test_data)
print('트레인 데이터',train_data.reshape(-1))
print('정규화 데이터',m_train_data.reshape(-1))
print('테스트 데이터',test_data.reshape(-1))
print('정규화 데이터',m_test_data.reshape(-1))
```
