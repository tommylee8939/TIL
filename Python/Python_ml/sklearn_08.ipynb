### iris clustering

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
iris_df
```

```python
# kmeans 객체를 생성하고 군집화 수행
iris_clustering = KMeans(n_clusters=3,random_state=3,verbose=1,init='k-means++',max_iter=300)
iris_clustering
```

```python
iris_clustering.fit(iris_df)
iris_clustering.labels_
```

```python
iris_clustering.predict(iris_df)
```

```python
iris_df['cid'] = iris_clustering.labels_
iris_df['target'] = iris.target
iris_df
```

```python
# c_id target을 통해 빈도를 확인해보자
iris_df.groupby(['cid','target'])['sepal length (cm)'].count()
```

```python
# 차원 축소를 위한 시각화 
from sklearn.decomposition import PCA
iris_pca = PCA(n_components=2)
iris_pca_transform = iris_pca.fit_transform(iris.data)

iris_df['pca_x'] = iris_pca_transform[:,0]
iris_df['pca_y'] = iris_pca_transform[:,1]
iris_df
```

```python
sns.lmplot('pca_x','pca_y',data=iris_df,fit_reg=False,scatter_kws ={'s':200},hue='cid')
```
