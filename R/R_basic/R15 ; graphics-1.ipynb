## plot( )
- generic function이다 
- 주어진 데이터 형식에 따라 가장 적합 형태의 그래프를 자동으로 출력한다 
- 벡터,팩터,데이터 프레임,시계열,선형회귀모델 ..
    - 두개의 벡터 => 산점도(scatter) 따로 scatter함수 없다
    - 벡터 (1 factor, 1 vector) => boxplot by factor
    - 데이터 프레임 (2개의 열) => 첫번째 열이 x , 두번째 열이 y
    - 데이터 프레임 3개이상의 열 => 산점도 행렬
    - 시계열 데이터 => time series plot 시계열 도표
    - 회귀분석모델 => 진단도표 diagnostic plot
    

```r
# 두개의 벡터 => 산점도(scatter) 따로 scatter함수 없다
plot(cars$speed,cars$dist)
```

```r
# 데이터 프레임 (2개의 열) => 첫번째 열이 x , 두번째 열이 y
plot(cars)
```

```r
# 벡터 (1 factor, 1 vector) => boxplot by factor
plot(iris$Species,iris$Sepal.Length)
```

```r
# 데이터 프레임 3개이상의 열 => 산점도 행렬
plot(iris[-5])
```

```r
# 시계열 데이터 => time series plot 시계열 도표
str(nhtemp)
plot(nhtemp)
```

```r
# table형식
str(UCBAdmissions)
UCBAdmissions
plot(UCBAdmissions)
```

```r
#회귀분석모델 => 진단도표 diagnostic plot
faithful.lm<-lm(waiting~eruptions,data=faithful)
plot(faithful.lm)
```

## pch
- plotting character type 즉 심볼 모양을 의미한다
- 기본값은 1
- 0에서 25번까지 각 정수에 대응하는 심볼로 데이터가 표현된다

## lty

- 그밖에 dahed longdash twodash dotdash 여러 옷변 존재

## type
- 특정 그래픽 함수들은 인자로 받는 데이터의구조에 따라 generic 타입에 맞게 type을 자동으로 지정해 준다
- type을 사용자가 직접 강제로 지정할 수도 있다

```r
#plot 데이터 프레임을 인자로 받으면 산점도 자동으로 지정한다
par(mfrow=c(2,2))
plot(pressure)
plot(pressure,type='l') # line
plot(pressure,type='o') # overplotted
plot(pressure,type='b') # both line and plot
```

```r
# 특별한 타입 n
# 고수준 그래픽을 먼저 그리고 단계별로 여러그래프를 그려나갈때 유용하다
# grid 레이어를 삽입하기 위해서도 사용한다 
par(mfrow=c(2,2))
x <- 1:10
y1<-exp(1:10)
y2<-exp(10:1)
plot(x,y1,type='n')
lines(x,y1,type='b',col='red')
lines(x,y2,type='b',col='blue')

plot(faithful,type='n')
grid()
points(faithful,col='blue')

plot(faithful) 
grid()        #그리드가 그래프를 부분적으로 덮어버리는 상황이 생길수 있다
```

## cex
- cex : character expression ratio
- 심볼이나 각종 텍스트 크리를 변경하기 위한 옵션이다 
- 상대적인 ratio다 => 1.5는 50%증가 0.5는 50% 감소를 의마한다

```r
par(mfrow=c(2,2))
x<-seq(0.5,1.5,0.25)
y<-rep(1,length(x))
plot(x,y,pch=19,main='main',xlab='xlab')
text(x,y+0.2,labels=x)

plot(x,y,cex=x,pch=19,main='main',xlab='xlab') # symbol 크기 변화
text(x,y+0.2,labels=x,cex=x) #텍스트 크기 변화

plot(x,y,cex=x,pch=19,main='main',xlab='xlab',cex.main = 1.5,cex.lab=1.5) # symbol + main + xlab 크기 변화
text(x,y+0.2,labels=x,cex=x)
```

## xlim ylim 
- x축과 y축긔 범위를 설정
- range함수를 이용해서 보다 정확하게 범위 설정하자
- 첫번째 데이터 셋 기준으로 창이 만들어지기 때문에 그래프가 다 안들어오는 상황 발생가능

```r
par(mfrow=c(2,2))
x1<-1:10
x2<-1:10
y1<-log(x1)
y2<-sqrt(x2)
plot(x1,y1,type='l',col='red')
lines(x2,y2,col='blue')

# 두 데이터 모두 가능한 범위 설정
xlim<-range(c(x1,x2))
ylim<-range(c(y1,y2))
plot(x1,y1,type='l',col='red',xlim=xlim,ylim=ylim)
lines(x2,y2,col='blue')
```

## layout 
- 좀더 다양한 형태 배치를 도와준다
- matrix를 인자로 넣어야한다

```r
# matrix를 먼저 살펴보자 
matrix(c(1,1,4,2,3,4),2,3,byrow=T)
# (1,1)(1,2)에 1번 그래프
# (1,3)(2,3)에 4번 그래프 
# (2,1) 에 2번 그래프
# (2,2) 에 3번 그래프 
# 이런식으로 옵션을 줄거다
```

```r
#widths heights 옵션
layout(matrix(c(1,1,4,2,3,4),2,3,byrow=T),
      widths=c(1,1,1.5),heights=c(2,1)) #widths 열의 폭, heights 행의 폭 을 조정해준다
plot(faithful)
plot(pressure)
plot(LakeHuron)
plot(iris$Species,iris$Sepal.Length)
```

## axis()
- side : 위치를 지정 (1,2,3,4) 아래쪽에서 시계방향
- at : 축 눈끔이 그려질 위치
- labels : 축 눈금의 (이름 생략하면 at 그대로 사용)
- tck : 눈금의 길이를 비율

```r
par(mfrow=c(2,2)) #공간을 만들어주자
states<-data.frame(state.x77)

# 라벨들 없애기 축 둘다 없애기
plot(states$Murder,states$Life.Exp,ylim=c(35,75),axes=F,ann=F)
# 라벨 없애고 yaxis 없애기
plot(states$Murder,states$Life.Exp,ylim=c(35,75),yaxt='n',ann=F)

plot(states$Murder,states$Life.Exp,ylim=c(35,75),yaxt='n',ann=F)
axis(side=4,at=seq(60,80,5),labels=seq(60,80,5),las=1)
plot(states$Murder,states$Life.Exp,ylim=c(35,75),yaxt='n',ann=F)
axis(side=4,at=seq(60,80,5),labels=seq(60,80,5),las=3)
```

## mtext(text,side,line)
- 그래프 마진 영역에 텍스트를 추가한다

```r
par(mar=c(4,8,4,8))
plot(states$Murder,states$Life.Exp,
     ylim=c(35,75),col='red',yaxt='n',ann=F)
points(states$Murder,states$HS.Grad,pch=22,col='blue',bg='skyblue')
#축 두개를 추가할거다
axis(side=2,at=seq(68,76,2),labels=seq(68,76,2),col.axis="red",las=2)
axis(side=4,at=seq(35,70,5),labels=seq(35,70,5),col.axis="blue",las=2)
mtext(text='High\nschool\nGraduate\nPercent',side=4,line=3,las=2,col='blue')
mtext(text='Life\nExpectation',side=2,line=3,las=2,col='red')
title(main='Murder vs Life Expectancy,\nMurder vs High School Graduates',
     xlab='Murder')
```

## legend(x,y,inset,title,legend,col,pch,lty,,,)
- x,y : 범례가 놓일 위치를 지정 =>bottom bottomleft top topleft left center 등으로 대체 가능하다 
- inset : 지정된 위치의 플롯 영역의 경계 사이에 삽입될 거리를 비율로 조정
- legend : 범례 character vector
- col pch lty lwd ... 실제 데이터가 구분되어있는 기준에 따라 벡터 값
- bg는 따로 설정 불가능한거 같다
- fill : 색상이 칠해진 면 막대,원 등
- 더 다양한 옵션 존재 text.col , horiz=TRUE

```r
tree1 <- subset(Orange,Tree==1)
tree2 <- subset(Orange,Tree==2)
xlim <- range(c(tree1$age,tree2$age))
ylim <- range(c(tree1$circumference,tree2$circumference))
# type both로 lty pch 다 있다
plot(tree1$age,tree1$circumference,type='b',xlim=xlim,ylim = ylim,
    pch=22,lty=1,col='red',bg='green')
lines(tree2$age,tree2$circumference,type='b',
      pch=24,lty=2,col='blue',bg='yellow')
legend('topleft',inset=0.05,title='Tree ID',legend=c('tree1','tree2'),
      lty=c(1,2),pch=c(22,24),col=c('red','blue'))
```

## text(x,y,labels,pos,offset,,,)
- mtext와 다르다
- 좌표값에 텍스트 배치
- labels에 실제 출력될 text를 문자 벡터로
- pos 아래쪽으로부터 시계방향으로 점 주변 위치 설정
- offset 점으로 부터 얼마나 떨어질지 설정
- 추가적으로 font cex col srt(각도)설정가능

```r
plot(1:3,1:3,type='n',ann=F)
text(1,1,labels="(1,1)",col="blue")
text(2,2,labels="(1,1)",col="blue",cex=4,srt=25)
text(2,2,labels="(1,1)",col="blue",cex=4,srt=25)
```

```r
plot(mtcars$wt,mtcars$mpg,pch=19)
text(mtcars$wt,mtcars$mpg,labels=row.names(mtcars),cex=0.6,pos=4)
```
