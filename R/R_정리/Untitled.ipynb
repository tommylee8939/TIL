#### 저수준 그래픽 함수를 호출하기 전에는 반드시 고수준 함수가 호출되어야 한다

```r
str(faithful)
```

```r
plot(faithful)
eruptions.big <- with(faithful,faithful[eruptions>3.5,])
points(eruptions.big,col='red',pch=19) # pch 19자체가 속이 채워진 원이다
```

```r
faithful.lm<-lm(waiting~eruptions,data=faithful)
plot(faithful)
eruptions.big <- with(faithful,faithful[eruptions>3.5,])
points(eruptions.big,col='red',pch=19)
lines(x=faithful$eruptions,fitted(faithful.lm),col='blue')
abline(v=3,col='purple') #vertical line
abline(h=70,col='green')
```

## plot( )
- generic function이다 
- 주어진 데이터 형식에 따라 가장 적합 형태의 그래프를 자동으로 출력한다 
- 벡터,팩터,데이터 프레임,시계열,선형회귀모델 ..
    - 두개의 벡터 => 산점도(scatter) 따로 scatter함수 없다
    - 벡터 (1 factor, 1 vector) => boxplot by factor
    - 데이터 프레임 (2개의 열) => 첫번째 열이 x , 두번째 열이 y
    - 데이터 프레임 3개이상의 열 => 산점도 행렬
    - 시계열 데이터 => time series plot 시계열 도표
    - 회귀분석모델 => 진단도표 diagnostic plot
    

```r
# 두개의 벡터 => 산점도(scatter) 따로 scatter함수 없다
plot(cars$speed,cars$dist)
```

```r
# 데이터 프레임 (2개의 열) => 첫번째 열이 x , 두번째 열이 y
plot(cars)
```

```r
# 벡터 (1 factor, 1 vector) => boxplot by factor
plot(iris$Species,iris$Sepal.Length)
```

```r
# 데이터 프레임 3개이상의 열 => 산점도 행렬
plot(iris[-5])
```

```r
# 시계열 데이터 => time series plot 시계열 도표
str(nhtemp)
plot(nhtemp)
```

```r
# table형식
str(UCBAdmissions)
UCBAdmissions
plot(UCBAdmissions)
```

```r
#회귀분석모델 => 진단도표 diagnostic plot
faithful.lm<-lm(waiting~eruptions,data=faithful)
plot(faithful.lm)
```

## 그래피 패러미터 설정

#### main, sub, xlab, ylab

```r
plot(faithful,
    main='title',
    sub='sub_title',
    xlab='xlab',
    ylab='ylab')
```

#### las(label style) 
#### bty(box type)
-위 두개는 그렇게 중요하지 않다

```r
par(mfrow = c(2,2))
plot(faithful,las=0)
plot(faithful,las=1)
plot(faithful,las=2)
plot(faithful,las=3)
```

```r
par(mfrow=c(2,2))
plot(faithful,bty='o')
plot(faithful,bty='n')
plot(faithful,bty='l')
plot(faithful,bty=']')
```
