#### 저수준 그래픽 함수를 호출하기 전에는 반드시 고수준 함수가 호출되어야 한다

```r
str(faithful)
```

```r
plot(faithful)
eruptions.big <- with(faithful,faithful[eruptions>3.5,])
points(eruptions.big,col='red',pch=19) # pch 19자체가 속이 채워진 원이다
```

```r
faithful.lm<-lm(waiting~eruptions,data=faithful)
plot(faithful)
eruptions.big <- with(faithful,faithful[eruptions>3.5,])
points(eruptions.big,col='red',pch=19)
lines(x=faithful$eruptions,fitted(faithful.lm),col='blue')
abline(v=3,col='purple') #vertical line
abline(h=70,col='green')
```

## plot( )
- generic function이다 
- 주어진 데이터 형식에 따라 가장 적합 형태의 그래프를 자동으로 출력한다 
- 벡터,팩터,데이터 프레임,시계열,선형회귀모델 ..
    - 두개의 벡터 => 산점도(scatter) 따로 scatter함수 없다
    - 벡터 (1 factor, 1 vector) => boxplot by factor
    - 데이터 프레임 (2개의 열) => 첫번째 열이 x , 두번째 열이 y
    - 데이터 프레임 3개이상의 열 => 산점도 행렬
    - 시계열 데이터 => time series plot 시계열 도표
    - 회귀분석모델 => 진단도표 diagnostic plot
    

```r
# 두개의 벡터 => 산점도(scatter) 따로 scatter함수 없다
plot(cars$speed,cars$dist)
```

```r
# 데이터 프레임 (2개의 열) => 첫번째 열이 x , 두번째 열이 y
plot(cars)
```

```r
# 벡터 (1 factor, 1 vector) => boxplot by factor
plot(iris$Species,iris$Sepal.Length)
```

```r
# 데이터 프레임 3개이상의 열 => 산점도 행렬
plot(iris[-5])
```

```r
# 시계열 데이터 => time series plot 시계열 도표
str(nhtemp)
plot(nhtemp)
```

```r
# table형식
str(UCBAdmissions)
UCBAdmissions
plot(UCBAdmissions)
```

```r
#회귀분석모델 => 진단도표 diagnostic plot
faithful.lm<-lm(waiting~eruptions,data=faithful)
plot(faithful.lm)
```

# 그래피 패러미터 설정

## main, sub, xlab, ylab

```r
plot(faithful,
    main='title',
    sub='sub_title',
    xlab='xlab',
    ylab='ylab')
```

## las(label style) bty(box type)
- 위 두개는 그렇게 중요하지 않다

```r
par(mfrow = c(2,2))
plot(faithful,las=0)
plot(faithful,las=1)
plot(faithful,las=2)
plot(faithful,las=3)
```

```r
par(mfrow=c(2,2))
plot(faithful,bty='o')
plot(faithful,bty='n')
plot(faithful,bty='l')
plot(faithful,bty=']')
```

## pch
- plotting character type 즉 심볼 모양을 의미한다
- 기본값은 1
- 0에서 25번까지 각 정수에 대응하는 심볼로 데이터가 표현된다
- col 옵션 줄수 있다
- 21부터 25까지는 속이 비어있는 형태여서 col(겉) + bg(속) 가각 지정할수 있다

```r
par(mfrow=c(2,2))
plot(faithful,pch=3)
plot(faithful,pch=21,col='green',bg='blue')
```

## lty
- lty : line type 선도표를 그릴때 선의 유형을 지정
- 기본값은 solid
- 그밖에 dahed longdash twodash dotdash 여러 옷변 존재
- 선에도 col 옵션 줄 수 있다

```r
par(mfrow=c(2,2))
plot(LakeHuron)
plot(LakeHuron,lty='solid')
plot(LakeHuron,lty='dashed')
plot(LakeHuron,lty='dotdash',col='red')
```

## type
- 특정 그래픽 함수들은 인자로 받는 데이터의구조에 따라 generic 타입에 맞게 type을 자동으로 지정해 준다
- type을 사용자가 직접 강제로 지정할 수도 있다

```r
#plot 데이터 프레임을 인자로 받으면 산점도 자동으로 지정한다
par(mfrow=c(2,2))
plot(pressure)
plot(pressure,type='l') # line
plot(pressure,type='o') # overplotted
plot(pressure,type='b') # both line and plot
```

```r
# 특별한 타입 n
# 고수준 그래픽을 먼저 그리고 단계별로 여러그래프를 그려나갈때 유용하다
# grid 레이어를 삽입하기 위해서도 사용한다 
par(mfrow=c(2,2))
x <- 1:10
y1<-exp(1:10)
y2<-exp(10:1)
plot(x,y1,type='n')
lines(x,y1,type='b',col='red')
lines(x,y2,type='b',col='blue')

plot(faithful,type='n')
grid()
points(faithful,col='blue')

plot(faithful) 
grid()        #그리드가 그래프를 부분적으로 덮어버리는 상황이 생길수 있다
```
