### Sorting 
- 1.order : vecter (index 반환 )
- 2.sort  : vecter
- 3.orderBy : data.frame


#### 1.order


```r
iris.df <- iris
iris.df$Sepal.Length[order(iris.df$Sepal.Length)]
```

```r
# 인덱스 => 즉 row index를 이용해서 data.frame 자체를 정렬 가능하다\
iris.df[order(iris.df$Sepal.Length),]
```

```r
# 만약에 값에 na가 포함된 경우
x<-c(1,2,4,3,NA,10,9,NA)
order(x) #na가 자동적으로 맨 뒤로 가게끔 index 준다
x[order(x,na.last = F)]
```

#### 2.sort( )

```r
sort(iris$Sepal.Length) # 정렬시켜준다
```

```r
sort(iris$Sepal.Length,decreasing = T)
```

```r
# 만약에 값에 na가 포함된 경우
x<-c(1,2,4,3,NA,10,9,NA)
sort(x) #일반적으론 na가 출력이 되지 않는다
sort(x,na.last = T)
sort(x,na.last = F)
```

### Unique
- 1. duplicated( ) => 중복여부를 논리값으로
- 2. unique( ) => 중복이 제거된 subset 

#### 1. duplicated

```r
# duplicated
# 처음 등장한 원소 False 그다음부턴 다 True
x<-c(1,2,3,1,3,1,2,3,1)
duplicated(x)
x.unique <- x[!duplicated(x)]
x.unique
```

```r
# 데이터 프레임을 인자로 하면 행에 대한 중복 여부를 체크한다
# 행에 모든 값이 일치하면 false를 준다 
id <- c(100,200,100,200)
name<-c("이철수","이순신","이철수","이순신")
salary<-c(100000,200000,100000,1)
df01 <- data.frame(id,name,salary)
duplicated(df01)
which(duplicated(df01))
df01[!duplicated(df01),] # 3행 제외
```

#### 2. unique()

```r
x<-c(1,2,3,1,3,1,2,3,1)
unique(x)
```

```r
id <- c(100,200,100,200)
name<-c("이철수","이순신","이철수","이순신")
salary<-c(100000,200000,100000,1)
df01 <- data.frame(id,name,salary)
unique(df01) # 3행 제외한 subset
```

### 결측지 제거
- 1. na.omit()
- 2. complete.cases() 
- 두개 동일하다

```r
x<-c(1,2,NA,1,2)
na.omit(x)
x[complete.cases(x)]
```

```r
id <- c(100,200,100,200)
name<-c("이철수","이순신",NA,"이순신")
salary<-c(NA,200000,100000,1)
df01 <- data.frame(id,name,salary)
df01
is.na(df01)
na.omit(df01)
```

```r
id <- c(100,200,100,200)
name<-c("이철수","이순신",NA,"이순신")
salary<-c(NA,200000,100000,1)
df01 <- data.frame(id,name,salary)
df01
df01[complete.cases(df01),]
```
