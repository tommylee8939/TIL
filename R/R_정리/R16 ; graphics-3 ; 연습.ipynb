## R16 ; graphics-3 ; 앞서 배운 기본 내장 그래픽 연습해보자

```r
# 0~25까지 심볼 그래피를 그려보자 
# points pch 따라 group 할수 있는 속성을 사용함
num <- 0:25
rows<- rep(1:4,each=7)
cols<- rep(7:1,4)
rows<- rows[1:(length(rows)-2)]
cols<- cols[1:(length(cols)-2)]
plot(rows,cols,type='n',ann=F,xaxt = 'n',yaxt ='n',xlim = c(1,4.5))
points(rows,cols,pch=num,col=(num+2))
text(rows,cols,labels = num,offset=2,pos=4) 
```

```r
# Line type 종류도 한번 보자
plot(1:5,1:5,type='n',xaxt = 'n',yaxt = 'n',ann=F)
abline(-2,1,lty='solid',col=1)
abline(-1,1,lty='dashed',col=2)
abline(0,1,lty='dotted',col=3)
abline(1,1,lty='longdash',col=4)
abline(2,1,lty='dotdash',col=5)
abline(3,1,lty='twodash',col=6)
legend('topleft',legend<-c('solid','dashed','dotted','longdash','dotdash','twodash'),
       lty =c('solid','dashed','dotted','longdash','dotdash','twodash'),
       col=1:6,
       cex=0.5)
```

```r
#ToothGrowth 데이터를 이용해서
#비타민 복용량과 방법에 따라 기니피그의 치아 길이의 분포를 boxplot으로
#boxplot formula형식 이용하면 된다
#?ToothGrowth
str(ToothGrowth)
boxplot(len~supp*dose,data=ToothGrowth,col=c('orange','yellow'))
legend('topleft',inset=0.05,legend=c('Oragne Juice','Ascorbic Acid'),fill=c('orange','yellow'))
```

```r
# ecnomics의 데이터 이용해서
# 년도에 따른 Unemployement 의 변화
str(economics)
plot(economics[c('date','unemploy')],type = 'o',pch=19,cex=0.5)
lines(lowess(economics[c('date','unemploy')]),col='blue')
abline(lm(unemploy~date,data=economics),lty='twodash',col='red')
legend('topleft',inset=0.05,legend=c('local-smoothing','linear-regression'),col=c('blue','red'),lty=c('solid','twodash'))
title(main = 'Unemployment Trend',col.main='tomato')
```
