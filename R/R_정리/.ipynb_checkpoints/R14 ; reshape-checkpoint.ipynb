## reshape

- 데이터셋은 일반적으로 두개의 형태를 갖게됨
    - wide format 우리가 아는 일반적인 data frame
    - long format variable 과 value 열을 통해 값들을 row 단위로 길게

```r
library(reshape2)
```

#### reshape2 패키지를 활용해서 실습
- melt( ) : wide to long
- dcast( ) : long to wide

#### melt( )
- factor가 자동으로 식별자로 지정됨
- id.var 옵션으로 식별자 지정 가능
- 식별자 여러개 지정가능 c(,)

```r
smiths
str(smiths)
```

```r
melt(smiths) # using subject as id variables
```

```r
melt(smiths,id.var='height')
```

```r
tail(melt(airquality)) #데이터에 factor 없다
```

```r
head(melt(airquality,id.vars=c('Month','Day')))
```

```r
#melt 함수 이용할때 옵션을 이용해서 보다 명식적으로 지정해보자
aq.long<-melt(airquality,
     id.vars = c("Month","Day"),
     measure.var = c("Ozone","Solar.R","Wind"), #Temp 뺏다
     varaiable.name = "변수",
     value.name ="값"
    )
head(aq.long)
```

#### dcast( )
- formula 형식 사용 x~y
- x는 식별자 변수 y 측정변수
- 여러개면 + 기호 사용
- 아무 변수도 포함시키 싫으면 .
- ... 지정안된 모든 변수

```r
smiths.long <- melt(smiths) 
airquality.long <-melt(airquality,id.vars=c('Month','Day'))
```

```r
dcast(smiths.long,subject~...)
```

```r
head(dcast(airquality.long,Month+Day~...))
head(dcast(airquality.long,Month+Day~variable,value.var = 'value'))
```

```r
#dcast함수 사용시 하나의 셀에 여러개의 값이 충될된다면?
aq.long<-melt(airquality,id.vars=c('Month','Day'))
head(aq.long)
#식별자 Month랑 Day였는데 Month 만 사용한다면?
dcast(aq.long,Month~variable)
#fun.aggregate 옵션 이용
dcast(aq.long,Month~variable,fun.aggregate = mean,na.rm=T)
```

```r
aq.long<-melt(airquality,
     id.vars = c("Month","Day"),
     measure.var = c("Ozone","Solar.R","Wind"), #Temp 뺏다
     variable.name = "변수",
     value.name ="값"
    )

head(dcast(aq.long,Month+Day~...))
# head(dcast(aq.long,Month+Day~varaiable)) 에러 난다 나는 variable이 아닌 변수로 썻으니깐
head(dcast(aq.long,Month+Day~변수,value.var = '값'))
```
