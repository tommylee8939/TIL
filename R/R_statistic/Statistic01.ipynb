## pnorm qnorm dnorm rnorm 무슨 차이일까?

```r
# dnrom => pdf of normal distribution
# dnorm(X) X : 실제 확률변수를 의미한다

# 정규분포 밀도 함수 그리기
x.lattice <- seq(-4,4,0.01)
plot(x.lattice,dnorm(x.lattice))
```

```r
# pnrom은 p여서 좀 헷갈리긴 하지만 누적분포를 의미한다
# 0 <=pnorm(X)<= 1 이어야 하고
# 여기서 X는 확률변수가 맞지만 Quantile 백분위수에 더 가깝다
# 즉 pnorm(Z0.25) = 0.25 가 되기 때문
# 우리가 알고있는 누적 정규확률분포를 그려보자
x.lattice <- seq(-4,4,0.01)
plot(x.lattice,pnorm(x.lattice))
```

```r
# rnrom 은 말그대로 normal dist에서 랜덤하게 표본을 추출하는거다
# rnom(count,mean,sd)
rnorm(1)
x<-rnorm(100,0,1)
hist(x)

y<-rnorm(1000,0,1)
hist(y) #=> 정규분포에 조금더 가까워지겠지
```

```r
# qnorm은 백분위수를 돌려준다
# qnorm(X) 0 <=X<= 1
# Ex) qnorm(0.5,0,1) 이면 당연히 평균인 0을 돌려줄거다
qnorm(0.5,0,1)

x.lattice <-seq(0,1,0.001)

# qnorm을 통해서도 역으로 누적분포 당연히 그릴수 있다
plot(qnorm(x.lattice),x.lattice)
```

### Empirical distribution (ecdf)

- 어떤 sample이 있다고 가정을 해보자
- 우리는 이 sample의 populaion distribtuion은 바로 알수가 없다
- 그래서 보통 한 sample의 분포를 cumulative density function F(X)로 주는것이 더 일반적이다


#### ecdf는 어떻게 구하는걸까? 

```r
# 지금 rnorm으로 데이터를 뽑았지만 실제로는 이게 정규분포에서 온 sample인 사실을 모른다고 가정을 해보자
x<-rnorm(100,0,1)

# 1. 통계 이론적응로 접근
# F = 1/n*sum(I{Xi<=X}}
percentile<-c()
for(i in 1:length(x)){
    sum<-1/length(x)*sum(ifelse(x<=x[i],1,0))
    percentile<-append(percentile,sum)
}
plot(x,percentile) # 정규분포에서 추출했으니깐 n이 커지면 정규분포랑 비슷하게 출력할거다


# 2. ecdf 함수를 이용해서 percentile 쉽게 구할수  있다
percentile2<-ecdf(x)(x)
plot(x,percentile2)


# 여기서 하나 더 궁금한게 딱 정확하게 quantile 50%를 어떻게 계산할까? => quantile함수 이용
```

### Quantile도 알아두자

```r
x<-c(1,2,3)
quantile(x)
quantile(x,probs = c(0.5))
```

### QQplot
- 우선 말그대로 Quantile과 Quantile을 찍은 그래프다
- 좌표값들이 직선상의 위치하면 두 분포가 서로 유사하다고 말할수 있다
- 반드시 y=x 일 필요는 없다
- 찍힌 점들을 바탕으로 regression line을 그릴수 있다
- 만약 점들이 regression line위에 놓인다면 a+bx => a:location b:scale이 된다

### 일반적으로 qqnorm()사용
- 샘플의 정규성을 파악할수 있다
- qqline()사용 가능
- 기본 built in 함수드에는 qqnorm 밖에 없다

```r
#size가 작으면 애매하게 보일수 있다
x<-rnorm(1000)
y<-runif(1000)
par(mfrow=c(1,2))
qqnorm(x)
qqline(x)
qqnorm(y)
qqline(y)
```

### 내가 직접 qqplot을 그리고 싶을때는?
- 하나의 sample이 주저졌을대 theoritical distribution이랑 비교 =>qnorm과 같이 실제분포 이용
- 두개의 smaple이 주어졌을때 비교해보자 => quantile이용

```r
# 1.하나의 sample이 주저졌다

x<-runif(1000,0,1)
percentile<-ecdf(x)(x)
plot(qnorm(,x) # theoritical distribution 입장에서는 percentile만 알면 quantile 바로 구할수 있다

# 아래처럼 해도 크게 문제는 없을거 같다
qqplot(x=,y=)
```

```r
# 2.지금 두개의 sample이 주어졌다고 가정을해보자

sample1<-runif(1000)
sample2<-rnorm(1000)
percentile_x<-ecdf(sample1)(sample1) #이 percentile에 해당하는 y의 quantile을 찾아야한다
quantile_y<-quantile(sample2,p=percentile_x)
plot(quantile_y,sample1)

# 쉽게 qqplot사용가능
qqplot(sample2,sample1)
```

```r
x<-runif(1000,0,1)
percentile_x<-ecdf(x)(x)
y<-rnorm(1000)
plot(quantile(y,prob=percentile_x),x)
     
```
