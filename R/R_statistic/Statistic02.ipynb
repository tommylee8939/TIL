### QQplot 심화
- qqnorm 말고도 여러 확률분포와도 비교가능하다
- 그에 따른 qqline도 그릴수 있다 
- location Scale도 찾을수 있다

```r
# norm 이 아닌 theoritical dist랑 데이터를 어케 비교하지?
par(mfrow=c(1,3))
data = rnorm(100, mean=5, sd=2)
percentile <- ecdf(data)(data)
plot(qexp(percentile),data)
qqline(data,distribution = qexp)

# 간단하게 qqplot이용
qqplot(qexp(percentile),data)
qqline(data,distribution = qexp)

# 더 간단하게 ppoints이용
qqplot(qexp(ppoints(100)),data) #ppoint 써도 대충은 비슷하게 나올거다 pppoints()는 데이터 갯수만큼 0~1까지 간격이 일정한 값을 만들어준다
# qqplot()을 사용할때는 앞에서 plot이 쌍을 만들어줘야됬던거와 다르게 그냥 데이터 두개만 집어넣으면 지가 알아서 qqplot만들어준다
qqline(data,distribution = qexp)
```

### qqline(data,distribution=)
- R에 내장되어있는 확률분포들 사용 가능하다
- 문제는 default값으로 1th 3th quantile을 잇는 선이다
- 실제  location scale을 조금 추정하고 싶으면 regression 해야된다

```r
par(mfrow=c(1,1))
data = rnorm(100, mean=5, sd=2)
percentile <- ecdf(data)(data)
plot<-qqplot(qexp(percentile),data)
qqline(data,distribution = qexp,col="red")
qqline(data,distribution = qexp,probs = c(0.3,0.6),col="green")

x<-plot$x
x[is.infinite(x)]<-NA
y<-plot$y
y[is.infinite(y)]<-NA
b<-sd(y,na.rm = T)/sd(x,na.rm = T)
a<-mean(y,na.rm = T)-b*mean(x,na.rm = T)
abline(a,b,col="blue")


```

### Location Scale
- The simplest case of such a deviation is when the QQ-plot is a straight line but not the line y = x, as in Figure 3.3. This is an indication that the data do not originate from F, but come from another member of the location-scale family of F

- 직선이긴 한데 y=x 는 아니네 그럼 location scale family인가 부다 
- 그럼 location이랑 scale을 어케 추정하지?


```r
# lcoation sacale추정해보자 

data1<- rnorm(10000,10,100) # => 이미 우리는 location family알고있다
plot<-qqnorm(data1)

x<-plot$x
x[is.infinite(x)]<-NA
y<-plot$y
y[is.infinite(y)]<-NA
b<-sd(y,na.rm = T)/sd(x,na.rm = T)
a<-mean(y,na.rm = T)-b*mean(x,na.rm = T)
abline(a,b,col="blue")
a;b # 데이터 크기가 작으면 막 그렇게 정확하지도 않다

qqline(data1,col="red")


```

## Heaviness of tail
- heaviness of tail 이 qqplot에 어떻게 영향을 주는지 이해할 수 있어야한다

```r
x<-seq(0,1,0.001)
x2<-seq(0,1,0.001)
par(mfrow=c(1,3))
qqplot(qunif(x,1,2),qnorm(x2,0,1))
qqplot(qnorm(x,1,2),qt(x2,df=3))
qqplot(qlnorm(x2,0.1,0.4),qt(x,df=10))
```

```r
# plot(qlnorm(x2,0.1,0.4),qt(x,df=10)) => 내가 내린 결론 왼쪽에선 t분포가 heavier하고 오른쪽에선 lnorm이 heavier이 되는거다
# 이를 한번 확인해보자
par(mfrow=c(1,2))

x<-seq(0,1,0.001)
sample1<-qlnorm(x,0.1,0.4)
sample2<-qt(x,df=10)
plot<-qqplot(sample1,sample2,xlab = "lnrom",ylab="t",type="l")


x<-plot$x
x[is.infinite(x)]<-NA
y<-plot$y
y[is.infinite(y)]<-NA
b<-sd(y,na.rm = T)/sd(x,na.rm = T)
a<-mean(y,na.rm = T)-b*mean(x,na.rm = T)
abline(a,b,col="blue",lty=1)

quantile_1<-c(quantile(sample1,probs =0.25),quantile(sample2,probs =0.25))
quantile_2<-c(quantile(sample1,probs =0.75),quantile(sample2,probs =0.75))

b<-(quantile_2[2]-quantile_1[2])/(quantile_2[1]-quantile_1[1])
a<-quantile_1[2]-b*quantile_1[1]
abline(a,b,col="red",lty=2)

legend("topleft",lty=c(1,2),legend=c("regression line","qqline"),col=c("blue","red"),)


x.lattice<-seq(-10,10,0.01)
plot(x.lattice,dlnorm(x.lattice,0.1,0.4),col="red",xlim=c(-6,6),type="l")
lines(x.lattice,dt(x.lattice,df=10),type="l")

```
