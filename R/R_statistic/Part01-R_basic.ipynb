### Vector 랑 List는 다른건가?
- 가장 쉽게 vector는 one type이고 one dimentional data를 다룬다
- 반면에 List는 multiple type이고 multiple dimentinal data를 다룬다

```r
num1<-c(1,2,"2")
class(num1) # class는 character이 된다
class(num1[1]) # 이미 character로 저장이 되어잇을거다

num2<-list(1,2,"2")
num2[[1]] # [[]]리스트 접근할때는 이거 사용해야함
names(num2)<-c("a","b","c")
class(num2)
class(num2$a) # [[]] 제외하고 $를 사용해서 이름으로 접근도 가능하다
```

```r
# 그럼 벡터안에 matrix랑 data.frame을 동시에 넣으면 어떻게 될까?

matrix1<-matrix(1:8,nrow=4,ncol=2)
frame1<-data.frame(a=c(1,2),b=c(1,2))

all<-c(matrix1,frame1)
class(all) # 이때는 list형이라고 해주네

list1<-c(frame1) #그냥 데이터 프레임 하나만 벡터형태로 담으면? => list다
class(list1)
```

#### 3.Data frame
- 앞으로 우리가 가장 많이 보게 될 자료형이 바로 data.frame이다
- data frame은 행렬을 넘어서서 테이블 형태를 가지고 있다
- 컬럼단위로 여러가지 자료형이 들어갈수 있다
- $를 통한 컬럼단위의 접근이 용이하다, ["컬러명"]도 가능하다

```r
no<-1:4
name<-c("Apple","Banana","Peach","Berry")
price <- c(500,200,300,50)
qty <-c(5,2,7,9)
dt1<-data.frame(과일명=name,가격=price,수량=qty)

dt1
class(dt1["과일명"]) #data frame
class(dt1[1])      #data frame

class(dt1$과일명) # factor(vector)
class(dt1[[c("과일명","가격")]]) # factor(vector)
```

```r
# 위 데이터 프레임에 자료를 추가하고 싶다 3가지 방법 가능하다
# 아래 방법 모두 동일하다 
dt2<-cbind(dt1,원산지=c("ko","ko","jpn","china"))
dt2

dt3<-dt1
dt3$원산지<-c("ko","ko","jpn","china")
dt3

dt4<-dt1
dt4["원산지"]<-c("ko","ko","jpn","china")
dt4
```

#### data frame 심화
- 이제 데이터 프레임을 가지고 더 자세하게 뜯어보자


```r
# [row,colum] [column]
# 즉 comma가 있으면 행,렬이고 없으면 colum으로 인식한다
# comma앞두의 아무것도 없으면 전체를 다 지칭한다

no<-1:4
name<-c("Apple","Banana","Peach","Berry")
price <- c(500,200,300,50)
qty <-c(5,2,7,9)
dt1<-data.frame(과일명=name,가격=price,수량=qty)
 
dt1[1] # 첫번째 컬럼 가져온다 => Data frame
dt1["과일명"] # 변수가 과일명인 Data frame
dt1[1:3,] # 1에서 3번째 까지 컬람 => Data frame
dt1[c(1,3),]# same

dt1[1,1] # 1행 1열 가져온다 => vector
dt1[[1]] # 안으로 들어가서 첫번째 컬럼이라고 생각하자 => vector
dt1[["과일명"]] # same => vector
```

```r
# list에서도 유사하게 사용가능
list_a <- list("dt1"=dt1,"dt2"=dt1)
list_a[1]
class(list_a[1]) #이거 아직  list다 재밌다
class(list_a[[1]])
class(list_a[[1]][1])
class(list_a[[1]][[1]])
```

### Indexing 심화

```r
sample <- c(1,2,3,4,5)
sample==1
#논리값으로 내가 원하는 값에 접근이 가능하다
sample[sample==1] # => sample[T,F,F,F,F]가 되는거다
sample[sample>1]  # => sample[F,T,T,T,T]가 되는거지
#직접적으로 값이 아니라 내가 원하는 index가 필요하면 그때 which함수 사용하는 거다
which(sample>1)
```

### 문자열 가지고 놀기

```r
xx<-c("a","b","c","d")
paste(xx[1],xx[2],xx[3])
paste("Hello","World") # 자동적으로 띄어쓰기 한칸이 들어간다
paste("Hello","World",sep=" ") # separator 구분 가능
```

```r
xx<-c("a","b","c","d")
paste(xx,xx,sep=",") #위의 예시와는 다르게 이 친구는 결과값이 하나가 아니라 여러개로 인식한다
paste(xx,xx,collapse = ",") #결과값이 여러개일때 ","를 이용해서 한줄로 이어주라는 뜻이된다
```

```r
# paste이용해서 1st 2nd 3rd 4th .. 9th까지 출력해보자
paste(1:9,c("st","nd","rd",rep("th",6))) #자동적으로 붙을때 띄어쓴다
paste(1:9,c("st","nd","rd",rep("th",6)),sep="") #옵션으로 붙여버리자
paste(1:9,c("st","nd","rd",rep("th",6)),sep="",collapse=",") #이제는 ,로 한줄로 출력하자
```

### 결측지, Infinite 처리해보자

```r
#의도적으로 Sepal.Length 1:20에는 NA
#의도적으로 Sepal.Width 1:20에는 Inf 값을 집어넣었다

data <- iris
data$Sepal.Length[1:20]<-NA
data$Sepal.Width[1:20] <-Inf
head(data[c(1,2)])
```

### 결측지 or Inf 값이 정말 존재하는지 어케알까?
- sum(is.na(data))>0
- sum(is.infinite(data))>0

```r
sum(is.na(data))>0
sum(is.na(iris))>0

sum(is.infinite(data$Sepal.Width))>0
```

### 결측치 or infinite의 위치를 찾는 방법
- which(x,logic) x는 logical vector or array다


```r
# 결측지의 위치를 한번 찾아볼까
which(is.na(data$Sepal.Length)==T) #1:20
data$Sepal.Length[which(is.na(data$Sepal.Length)==T)] # 여기서 값을 변경해주면 되는거다

which(is.infinite(data$Sepal.Width))
# is.na => true or false값을 돌려줌
#sum(is.na(data$Sepal.Length)) # 0보다 크면 na값이 있는거다
```
